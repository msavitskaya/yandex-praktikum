## Замена пропусков в столбцах

Начнем с изучения пропусков: 
* пропуски в поле **ceiling_height** не могут быть, ведь у каждой квартиры и дома должен быть потолок, подсчитаем количество объявлений без указания этого параметра, если таких объявлений более 1%, то попробуем заменить пропуски на медиану по похожим объектам, иначе удалим
* пропуски в поле **floors_total** нужно будет изучить, возможно, это частные дома
* также есть пропуски в поле **living_area**, это важная характеристика квартир, поскольку мы рассматриваем жилые объекты, то жилая площадь должна быть
* пропуски в поле **kitchen_area**, также важная характеристика, в некоторых случаях её может не быть, например у квартир-студий
* почти в половине квартир есть пропуски в поле **balcony**, такое вполне может быть, у кого-то нет балконов, у кого-то лоджие вместо балконов, заполним эти пропуски нулями
* пропуски в поле **locality_name** очень смущают и помешают нам сделать правильные выводы, поступим с ними также как с полем ceiling_height, если меньше 1% удалим данные
* пропуски в поле **airports_nearest** и **cityCenters_nearest** посчитаны автоматически. Для airports_nearest, предположительно, в населенном пункте может не быть аэропорта, поэтому значения пустые. А для cityCenters_nearest возможно в данных яндекс.карт мы не смогли определить центр населенного пункта или объект очень сильно удален от центра

Жилую площадь можно найти вычитанием площади кухни от общей площади квартиры. У общей площади пропусков нет, проверим совпадают ли пропуски в жилой площади и площади кухни

Жилая площадь и площадь кухни больше всего зависит от общей площади. Заполним пропуски значениями в зависимости от общей площади

```python
# заполнение пропусков в столбце kitchen_area на 17% от величины общей площади
data['kitchen_area'] = data['kitchen_area'].fillna(data['total_area']*0.17)
print('Проверим, что пустых строк в столбце kitchen_area не осталось:',data['kitchen_area'].isna().sum())
```

Проверим, что теперь в данных площадей все цифры реальные, для этого посчитаем количество значений, где сумма жилой площади и кухни больше чем общая площадь

```python
print('Проверим сколько объектов имеют сумму кухни и жилой площадей больше чем общая площадь объекта:',len(data.query('kitchen_area+living_area > total_area')))
```

Очевидно в данных площадей где-то закралась ещё ошибка, наша замена не могла превысить общую площадь, значит эти 116 некорректных объекта были изначально. В данном случае проще удалить эти объекты из анализа, так как их немного, иначе мы много времени потратим на поиск ошибки.

```python
# удаление объектов, которые не удовлетворяют условию itchen_area+living_area <= total_area
data.drop(data[data['kitchen_area'] +  data['living_area'] > data['total_area']].index, inplace=True)
print('Проверим сколько объектов имеют сумму кухни и жилой площадей больше чем общая площадь объекта:',len(data.query('kitchen_area+living_area > total_area')))
```

Очень мало данных без наименования населенного пункта, в нашем задании важно понимать где находится объект, поэтому удалим эти значения, чтобы они не искажали последующий анализ данных.

```python
# удаление пропущенных значений в населенных пунктах
data.dropna(subset = ['locality_name'],inplace = True)
print('Проверим, что пустых строк в столбце locality_name не осталось:',data['locality_name'].isna().sum())
```

```python
list_name = ['городской поселок ',
             'поселок городского типа ',
            'коттеджный поселок ', 
            'поселок при железнодорожной станции ',
            'при железнодорожной станции ',
            'поселок станции ',
            'поселок ',
            'село ',
            'деревня ',
            'садоводческое некоммерческое товарищество ',
            'садовое товарищество ']
# функция чистки названия города от слов поселок, деревня и прочее
def clean_city_name(city):
    for index in range(len(list_name)):
        if list_name[index] in city:
            city_new = city.replace(list_name[index], '')
            return city_new
    return city
data['locality_name'] = data['locality_name'].apply(clean_city_name)
# вывод уникальных значений в столбце locality_name
print('Количество уникальных городов стало', len(data['locality_name'].unique()))
```

Попробуем найти значение расстояния до ближайшего аэропорта среди имеющихся данных по населенным пунктам, возможно в данных имеются такие же населенные пункты, у которых поле airports_nearest заполнено

```python
# функция для категоризации этажа квартиры; варианты — первый, последний, другой
def do_group_for_floor(row):
    if row['floor'] == 1:
        return 'Первый'
    elif row['floor'] == row['floors_total']:
        return 'Последний'
    return 'Другой'

# применение функции do_group_for_floor ко всему столбцу 'floor'
data['floor_group'] = data.apply(do_group_for_floor, axis = 1)
data.groupby('floor_group')['floor_group'].agg({'count'})
```

Двухкомнатные и трехкомнатные квартиры на втором месте, ну и затем 4х, 5ти и 6ти комнатные квартиры

В наших данных есть выбросы, квартиры сроком продажи более 464 дней, это самые долгие продажи по нашим данным, посмотрим сколько таких квартир

```python
# расчет первого квартиля
q1 = data.days_exposition.quantile(0.25)
# расчет третьего квартиля
q3 = data.days_exposition.quantile(0.75)
# расчет межквартильного размаха
iqr = q3 - q1
data_filtered_days_exposition = data.query('days_exposition <= @q3 + 1.5 * @iqr and days_exposition > 0')
print('График размаха усов для среза по количеству дней продажи: ( ',0, ',', q3 + 1.5 * iqr, ')')
# график размаха усов
data_filtered_days_exposition.boxplot('days_exposition')
```

```python
# Гистограмма распределения количества дней публикации
sns.histplot(data=data_filtered_days_exposition, x='days_exposition', kde=True)
# установим заголовок для графика
plt.title("Гистограмма распределения количества дней публикации"); 
# подпишем оси координат обозначив единицы измерения
plt.ylabel('Количество объявлений')
plt.xlabel('Kоличество дней публикации')
```

Чем больше дней прошло с момента публикации объявления, тем реже покупают квартиры, возможно дело в самом объявлении, нет картинок, плохое описание, большая удаленность от центра

### Анализ факторов влияющих на стоимость квартиры

* Зависит ли цена от общей площади
* Зависит ли цена от удалённости от центра
* Зависит ли цена от числа комнат
* Зависит ли цена от того, на каком этаже расположена квартира: первом, последнем или другом
* Зависит ли цена от даты размещения объявления: день недели, месяц, год

Построим диаграмму рассеяния для визуальной оценки взаимосвязи стоимости квартиры и общей площади

```python
# диаграмма рассеяния по столбцам 'цена' и 'общая площадь'
pd.plotting.scatter_matrix(data_filtered.loc[:,['last_price' , 'total_area']], diagonal='kde', figsize=(9, 9))
# коэффициент корреляции
corr = data_filtered['last_price'].corr(data_filtered['total_area'])
print('Коэффициент корреляции: {:.1f}'.format(corr))
```

Коэффициент корреляции Пирсона равен 0.7, это значит, что между параметрами цена и общая площадь имеется прямая зависимость, чем больше площадь, тем больше цена

Коэффициент корреляции Пирсона равен всего -0.3, это говорит о том, что между параметрами "цена" и "расстояние до центра" корреляция практически отсутствует. Это же демонстрируют нам графики, квартиры с удаленностью от центра от 10-20 км. имеют абсолютно разные цены от 2 млн.рублей до 10 млн.рублей

В этом проекте были изучены данные архива объявлений о продаже квартир в Санкт-Петербурге и соседних населенных пунктах за пять лет. В первом этапе работы мы провели анализ пропущенных значений и проверили предобработку данных. Заменили типы столбцов там где это было необходимо.

Во втором этапе, в таблицу добавили столбцы «Цена кв.метра», «День недели», «Название месяца» и «Год» публикации объявления. Кроме того, добавлен столбец «Статус этажа» (первый, последний, другой), рассчитаны соотношения жилой и общей площадей, а также площади кухни и общей площади.

На третьем этапе мы проверила анализ данных, в результате чего сделали следующие выводы:

* Большинство квартир в данных это однокомнатные квартиры с площадью 37 квадратных метров
* Самая маленькая квартира 13 квадратов, а самая большая квартира целых 900
* Чаще всего встречаются квартиры с высотой потолкой 2.65
* Одна из квартир продавалась более 4х лет(1580 дней)
* Дольше всего продаются квартиры большой комнатности от 8 и выше. Быстрее всего однокомнатные квартиры, средний срок экспозиции на сайте 143 дня. 
* Самые долгие продажи по нашим данным более 464 дней
* Большинство квартир продается в течение 100 дней с момента публикации объявления, при этом пик продаж приходится с 40-60 день
* Чем больше дней прошло с момента публикации объявления, тем реже покупают квартиры
* Чем больше площадь, тем больше цена
* На первом этаже квартиры дешевле всего, самые дорогие квартиры на не первом и не последнем этажах
* К лету есть небольшой тренд снижения цен, а зимние месяцы наоборот самые популярные и цены немного подрастают
* Стоимость жилой недвижимости за последние пять лет не сильно менялась, но наблюдаем рост в 2019 году, в отличие от Санкт-Петербурга, где квартиры в центре стали стоить на уровне 2015г
* Ниболее значимые факторы, которые больше всего влияют на стоимость квартиры: Количество комнат, Этаж квартиры, Площадь квартиры
* Населённый пункт с самой высокой стоимостью жилья: Санкт-Петербург = 103 тыс. рублей за м², населённый пункт с самой низкой стоимостью жилья: Выборг = 56 тыс. рублей за м²
