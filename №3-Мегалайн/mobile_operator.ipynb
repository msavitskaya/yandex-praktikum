# Определение перспективного тарифа для телеком компании

Клиентам предлагают два тарифных плана: «Смарт» и «Ультра». Чтобы скорректировать рекламный бюджет, коммерческий департамент хочет понять, какой тариф приносит больше денег.

**Тариф «Смарт»**
* Ежемесячная плата: 550 рублей
* Включено 500 минут разговора, 50 сообщений и 15 Гб интернет-трафика

Стоимость услуг сверх тарифного пакета:
* минута разговора: 3 рубля
* сообщение: 3 рубля
* 1 Гб интернет-трафика: 200 рублей

**Тариф «Ультра»**
* Ежемесячная плата: 1950 рублей
* Включено 3000 минут разговора, 1000 сообщений и 30 Гб интернет-трафика

Стоимость услуг сверх тарифного пакета:
* минута разговора: 1 рубль
* сообщение: 1 рубль
* 1 Гб интернет-трафика: 150 рублей

Необходимо проверить гипотезы: 
* средняя выручка пользователей тарифов «Ультра» и «Смарт» различается;
* средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов.

### Замена пропущенных значений в таблице users

В ходе знакомства с данными мы обнаружили пропущенные значения в таблице users в столбце churn_date, в дальнейшем анализе будем учитывать, что ноль - это те клиенты, которые на момент выгрузки еще пользовались тарифом.

```python
print('Доля пропущенных значений в столбце churn_date = {:.1%}'.format(users['churn_date'].isna().sum() / len(users)))
```

```python
# замена пропущенных значений на медиану
users['churn_date'] = users['churn_date'].fillna(0)
print('Проверим, что пустых строк в столбце churn_date не осталось:',users['churn_date'].isna().sum())
```

Наблюдаем пик количества звонков в нуле, скорее всего это дозвон и сброс, а возможно прерывание связи.

```python
print("Количество звонков длительностью 0 минут", calls.query('duration == 0')['duration'].count())
print("Доля звонков длительностью 0 минут от общего числа данных {:.1%}". format(calls.query('duration == 0')['duration'].count()/ calls.shape[0]))
```

```python
# удаление объектов с длительностью звонка 0 минут
calls.drop(calls[calls['duration']== 0].index, inplace=True)
print('Проверим, что нулевых значений в столбце duration не осталось:',calls.duration.isna().sum())
```

### Расчет выручки с каждого клиента

Посчитаем помесячную выручку с каждого пользователя по формуле:

Выручка за месяц = ((Сумма длительности звонков- Бесплатный лимит минут) + (Сумма сообщений - Бесплатный лимит сообщений) + (Сумма Мб - Бесплатный лимит Мб)) * Стоимость тарифа + Абонентская плата

```python
# функция подсчета стоимости минут сверхплана
# возвращает 0, если бесплатный лимит не исчерпан
# возвращает сверхлимит * на стоимость минуты по тарифу, если бесплатный лимит исчерпан
def count_calls_over_plan(row):
    if row['tariff'] == 'smart':
        result = np.ceil(row['duration_calls'] - 500)
        return np.where(result > 0, result*3, 0)
    else:
        result = np.ceil(row['duration_calls'] - 3000)
        return np.where(result > 0, result, 0)
# добавление столбца с суммой доплаты по тарифу
all_info['cost_minutes_over'] = all_info.apply(count_calls_over_plan, axis=1)
all_info
```

```python
# функция подсчета стоимости сообщений сверхплана
# возвращает 0, если бесплатный лимит не исчерпан
# возвращает сверхлимит * на стоимость смс по тарифу, если бесплатный лимит исчерпан
def count_messages_over_plan(row):
    if row['tariff'] == 'smart':
        result = row['count_messages'] - 50
        return np.where(result > 0, result*3, 0)
    else:
        result = row['count_messages'] - 1000
        return np.where(result > 0, result, 0)
# добавление столбца с суммой доплаты по тарифу
all_info['cost_sms_over'] = all_info.apply(count_messages_over_plan, axis=1)
all_info
```

```python
# функция подсчета стоимости Мб сверхплана
# возвращает 0, если бесплатный лимит не исчерпан
# возвращает сверхлимит * на стоимость Гб по тарифу, если бесплатный лимит исчерпан
def count_mb_over_plan(row):
    if row['tariff'] == 'smart':
        result = np.ceil(row['sum_mb'] / 1024 - 15)
        return np.where(result > 0, result * 200, 0)
    else:
        result = np.ceil(row['sum_mb'] / 1024 - 30)
        return np.where(result > 0, result * 150, 0)
# добавление столбца с суммой доплаты по тарифу
all_info['cost_mb_over'] = all_info.apply(count_mb_over_plan, axis=1)
all_info
```

```python
# функция подсчета выручки по каждому тарифу
def count_payment(row):
    if row['tariff'] == 'smart':
        sum = row['cost_minutes_over'] + row['cost_sms_over'] + row['cost_mb_over'] + 550
    else:
        sum = row['cost_minutes_over'] + row['cost_sms_over'] + row['cost_mb_over'] + 1950
    return sum    
# добавление столбца с выручкой в рублях в таблицу
all_info['payment'] = all_info.apply(count_payment, axis=1)
all_info.sample(10)
```

```python
# выручка по тарифному плану smart
smart_payment = all_info.query('tariff =="smart"').groupby(by='month').agg({'payment':'sum'}).reset_index()
# выручка по тарифному плану ultra
ultra_payment = all_info.query('tariff =="ultra"').groupby(by='month').agg({'payment':'sum'}).reset_index()
# график выручки по месяцам
ax = (smart_payment\
     .plot(y='payment', x='month', kind='bar', alpha=0.3, grid=True, legend=True, color='r', label = 'Тариф Smart'))
ultra_payment\
     .plot(y='payment', x='month', kind='bar', alpha=0.5, grid=True,  ax=ax, legend=True, label = 'Тариф Ultra', figsize=(10,5))
plt.title('Выручка по месяцам')
plt.xlabel('Месяц')
plt.ylabel('Размер выручки')
plt.show()
```

Интересно, что с каждым месяцем по обоим тарифам наблюдается тенденция роста, возможно каждый месяц значительно увеличивалась доля клиентов оператора. # Шаг 4. Проверка гипотез <a id="step4"></a>

### Гипотеза 1: средняя выручка пользователей тарифов «Ультра» и «Смарт» различается

H0 (нулевая гипотеза): средняя выручка пользователей тарифов «Ультра» и «Смарт» не различается.

H1 (альтернативная гипотеза): средняя выручка пользователей тарифов «Ультра» и «Смарт» различается.

Для проверки гипотезы воспользуемся методом scipy.stats.ttest_ind, так как выборки независимы.

```python
#  выручка для тарифа smart
smart_payment = all_info.query('tariff == "smart"')['payment']

# гистограмма распределения выручки smart
sns.histplot(data=smart_payment, kde=True)
# установим заголовок для графика
plt.title("Гистограмма распределения выручки smart"); 
# подпишем оси координат обозначив единицы измерения
plt.ylabel('Частота')
plt.xlabel('Размер выручки')
plt.figure(figsize=(10,7))

print('Дисперсия выборки тарифа smart = {} '.format(np.var(smart_payment, ddof=1)))
```

```python
# выручка для тарифа ultra
ultra_payment = all_info.query('tariff == "ultra"')['payment']

# гистограмма распределения выручки ultra
sns.histplot(data=ultra_payment, kde=True)
# установим заголовок для графика
plt.title("Гистограмма распределения выручки ultra"); 
# подпишем оси координат обозначив единицы измерения
plt.ylabel('Частота')
plt.xlabel('Размер выручки')
plt.figure(figsize=(10,7))
print('Дисперсия выборки тарифа smart = {} '.format(np.var(ultra_payment, ddof=1)))
```

```python
# критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу
alpha = .05 
results = st.ttest_ind( smart_payment, ultra_payment,equal_var = False)
print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Средняя выручка пользователей тарифов «Ультра» и «Смарт» различается. Что ожидаемо, так как абоненсткая плата по тарифу «Ультра» в три раза больше чем по тарифу «Смарт».

### Гипотеза 2: средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов

H0 (нулевая гипотеза): средняя выручка пользователей из Москвы не отличается от выручки пользователей из других регионов

H1 (альтернативная гипотеза): средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов

Для проверки гипотезы воспользуемся методом scipy.stats.ttest_ind, так как выборки независимы.

```python
#  выручка клиентов из Москвы 
moscow = all_info.query('city =="Москва"')['payment']

# гистограмма распределения выручки Москвы
sns.histplot(data=moscow, kde=True)
# установим заголовок для графика
plt.title("Гистограмма распределения выручки Москвы"); 
# подпишем оси координат обозначив единицы измерения
plt.ylabel('Частота')
plt.xlabel('Размер выручки')
plt.figure(figsize=(10,7))
print('Дисперсия выборки выручки по Москве = {} '.format(np.var(moscow, ddof=1)))
```

```python
#  выручка клиентов из других городов
other_district = all_info.query('city != "Москва"')['payment']

# гистограмма распределения выручки других городов
sns.histplot(data=other_district, kde=True)
# установим заголовок для графика
plt.title("Гистограмма распределения выручки других городов"); 
# подпишем оси координат обозначив единицы измерения
plt.ylabel('Частота')
plt.xlabel('Размер выручки')
plt.figure(figsize=(10,7))
print('Дисперсия выборки выручки по Москве = {} '.format(np.var(other_district, ddof=1)))
```

```python
# критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу
alpha = .05 
results = st.ttest_ind( moscow, other_district,equal_var = False)
print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Москва большой город, но тем не менее в среднем выручка пользователей не отличается от выручки пользователей из других регионов.
