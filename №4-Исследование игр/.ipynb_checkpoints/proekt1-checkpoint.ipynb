Рассмотрим данные в user_score.

С оценками пользователей та же беда, что и с оценками критиков, думаю эти игры также просто не оценивались, либо оценок очень мало.

```python
# уникальные значения столбца
games.user_score.unique()
```

В данных есть значение tbd - это значит To Be Determined, то есть оценка пользователями ещё не дана, возможно игра непопулярная или вышла недавно. В данном случае можно было бы попробовать предсказать оценку пользователей по оценке критиков, но значения находятся в разных шкалах измерения, оценка критиков от 0-100, пользователей от 0-10, к тому же в оценке критиков очень много пропусков, как мы выяснили выше больше 50%. Следователно, предлагаю поступить также как со столбцом critic_score, заменим пропуски и tbd на -1, что будет означать, что оценки пользователей нет у игры.

```python
# замена пропусков user_score на -1
games['user_score'] = games['user_score'].fillna(value=-1)
# замена tbd на -1
games.user_score = games.user_score.replace('tbd', -1)
print('Проверим, что пустых строк и значений "tbd" в столбце user_score не осталось:',\
      games['user_score'].isna().sum()+games.query('user_score == "tbd"').user_score.count())
```

```python
# приведение к типу float
games.user_score = games.user_score.astype('float')
print('Привели столбец user_score к типу',games.user_score.dtypes) 
```

### Столбец rating

Рассмотрим данные в rating.

```python
 # уникальные значения столбца
games.rating.unique()
```

Изучим данные предметной области, что значат эти буквы в рейтингах компьютерных игр. В данных представлен рейтинг негосударственной организации Entertainment Software Rating Board (ESRB), которая определяет возрастной рейтинг компьютерных игр.
1. «E» («Everyone») — «Для всех»: Содержание вполне подходит для возрастной категории от 6 лет и старше. Первоначально "K-A" ("Kids to Adults").
2. «M» («Mature») — «Для взрослых»: Материалы игры не подходят для подростков младше 17 лет. 
1. «T» («Teen») — «Подросткам»: Игра подходит для лиц от 13 лет и старше.
1. «E10+» («Everyone 10 and older») — «Для всех от 10 лет и старше».
1. «AO» («Adults Only 18+») — «Только для взрослых».Содержание игры только для взрослых старше 18 лет. 
1. «EC» («Early childhood») — «Для детей младшего возраста»: Игра подходит для детей от 3 лет и старше
1. «RP» («Rating Pending») — «Рейтинг ожидается»: Продукт был отправлен в ESRB и ожидает присвоения рейтинга.

Оказывается рейтинг "K-A" и «E» это одно и тоже, поэтому смело можем объединить значения в одну группу, допустим «E».

```python
# замена типа "K-A" на «E»
games.rating = games.rating.replace('K-A', 'E')
 # уникальные значения столбца
print('Заменили рейтинг "K-A" на "E", теперь остались только следующие рейтинги: \n', games.rating.unique())
```

```python
print('Количество пропущенных значений в столбце rating =',len(games[games['rating'].isnull() == True]))
print('Доля пропущенных значений составляет = {:.1%}'.format(len(games[games['rating'].isnull() == True]) / games.shape[0]))
```

Очень большое количество разных игр, наверняка компания ESRB дает рейтинг только самым продаваемым или популярным, поэтому не удивительно, что у каких-то игр нет этого рейтинга.

Очень много пропущенных значений, заменим пустые строки в столбце rating на "RP", что значит что рейтинг ожидается, то есть не определен ещё.

```python
# замена пропусков rating на RP
games['rating'] = games['rating'].fillna(value='RP')
print('Проверим, что пустых строк в столбце rating не осталось:',games['rating'].isna().sum())
```

### Посмотрим, сколько игр выпускалось в разные годы

```python
# таблица с количеством игр в разные годы
count_games_for_year = games.pivot_table(index='year_of_release', values='name', aggfunc='count').query('year_of_release > 0').reset_index()
# график с количеством игр в разные годы
sns.set_theme(style="whitegrid")
sns.catplot(\
            kind="bar", data=count_games_for_year,  x='year_of_release', y='name', \
            alpha=.6, color='#008000', height=6, aspect=2, orient='v'\
           )
# установим заголовок для графика
plt.title("График количества игр по году выпуска"); 
# подпишем оси координат обозначив единицы измерения
plt.ylabel('Количество игр')
plt.xlabel('Год')
plt.xticks(rotation=30)
plt.show()
```

### Посмотрим, как менялись продажи по платформам

```python
# функция создания объема продаж по заданному столбцу 
def find_sales(data,columns,values):
    pivot_table = data\
                        .pivot_table(index=columns, values=values, aggfunc='sum')\
                        .sort_values(by = values,ascending = False)\
                        .reset_index()
    return pivot_table
# создание таблицы объема продаж по платформам
count_sales_for_platform = find_sales(games,'platform','sales')

# список порядка для вывода категорий платформ на оси х
x_order = count_sales_for_platform['platform'].to_list()

# график с количеством игр в разные годы
sns.set_theme(style="whitegrid")
sns.catplot(
            kind="bar", data=count_sales_for_platform,  x='platform', y='sales', \
            alpha=.6, color='#008000', height=6, aspect=2, orient='v', order=x_order\
             )
# установим заголовок для графика
plt.title("График продаж по платформам"); 
# подпишем оси координат обозначив единицы измерения
plt.ylabel('Сумма продаж, млн.')
plt.xlabel('Платформа')
plt.xticks(rotation=30)
plt.show()
```

На графике видно, что платформы PS2, X360, PS3, WII, DS, PS значително превосходят другие платформы по мировым продажам, ну они и самые популярные на мой взгляд.

Далее посмотрим график распеределения продаж по годам на этих платформах.

```python
# функция создания таблицы с ежегодной суммой продаж по годам
def make_pivot_table_platform(platform_name):
    pivot_table = games.query('platform == @platform_name and year_of_release>=1995')\
                        .pivot_table(index='year_of_release', values='sales', aggfunc='sum')\
                        .reset_index()
    pivot_table['platform'] = platform_name
    return pivot_table
    
# таблицы с суммой всех продаж по годам
count_sales_for_ps2 = make_pivot_table_platform("ps2")
count_sales_for_x360 = make_pivot_table_platform("x360") 
count_sales_for_ps3 = make_pivot_table_platform("ps3") 
count_sales_for_wii = make_pivot_table_platform("wii") 
count_sales_for_ds = make_pivot_table_platform("ds") 

# создание итоговой таблицы со всеми данными для построения графика
top5_sales = count_sales_for_ps2\
                .append(count_sales_for_x360)\
                .append(count_sales_for_ps3)\
                .append(count_sales_for_wii)\
                .append(count_sales_for_ds)
# график с продажами по годам
sns.set_theme(style="whitegrid")
sns.catplot(
            kind="point", data=top5_sales,  x='year_of_release', y='sales', hue="platform", \
            alpha=.6, height=6, aspect=2, orient='v'\
            )
# установим заголовок для графика
plt.title("График ежегодных продаж для топ5 платформ"); 
# подпишем оси координат обозначив единицы измерения
plt.ylabel('Сумма продаж, млн.')
plt.xlabel('Года')
plt.xticks(rotation=30)
plt.show()
```

### Определим для пользователя каждого региона (NA, EU, JP) самые популярные платформы (топ-5)

```python
# функция создания таблицы с продажами по столбцу в разных регионах    
def create_table_top5_region(column):
    # создание таблиц по объему продаж по столбцу column в регионах NA, EU, JP
    count_sales_in_na = find_sales(games_new,column,'na_sales').loc[:4]
    count_sales_in_eu = find_sales(games_new,column,'eu_sales').loc[:4]
    count_sales_in_jp = find_sales(games_new,column,'jp_sales').loc[:4]

    # добавление столбца с регионом для орисовки на графике
    count_sales_in_na['region'] = 'North America'
    count_sales_in_eu['region'] = 'Europe'
    count_sales_in_jp['region'] = 'Japan'

    # переименование столбца na_sales, eu_sales, jp_sales для будущего объединения таблиц
    count_sales_in_na = count_sales_in_na.rename(columns={'na_sales': 'sales'})
    count_sales_in_eu = count_sales_in_eu.rename(columns={'eu_sales': 'sales'})
    count_sales_in_jp = count_sales_in_jp.rename(columns={'jp_sales': 'sales'})

    # объединение таблиц 
    count_sales_in_region = count_sales_in_na\
                .append(count_sales_in_eu)\
                .append(count_sales_in_jp)\
                .reset_index()
    return count_sales_in_region

# таблица с продажами по платформе в разных регионах    
count_sales_for_platform_in_region = create_table_top5_region('platform')

# определение платформ для отрисоки на оси х                        
x_order = count_sales_for_platform_in_region['platform'].unique()

# график с продажами по платформам в разных регионах
sns.set_theme(style="whitegrid")
sns.catplot(
            kind="bar", data=count_sales_for_platform_in_region,  x='platform', y='sales', hue="region", \
            alpha=1, height=6, aspect=2, orient='v', order = x_order\
            )
# установим заголовок для графика
plt.title("График ежегодных продаж для топ5 платформ в разных регионах"); 
# подпишем оси координат обозначив единицы измерения
plt.ylabel('Сумма продаж, млн.')
plt.xlabel('Платформы')
plt.xticks(rotation=30)
plt.show()
```

Самые популярные платформы в Северной Америке это x360, ps2, wii, ps3, ds. Те же самые платформы мы получили при анализе мировых продаж. Только в Северной Америке платформа x360 оказалась более прибыльна, чем платформа ps2.

Самые популярные платформы в Европе это ps3, x360, wii ,ds, ps4. А в Европе платформа x360 немного проигрывает по продажам платформе ps3. Но в целом картинка повторятся, тот же топ-5 платформ, что и на мировом рынке.

В Японии топ-5 популярных платформ занимают ds, 3ds, ps3, psp, wii. 

Посчитаем долю продаж для каждой платформы.

```python
# функция подсчета процента продаж
def count_percent(data,index):
    for column in ['na_sales', 'eu_sales', 'jp_sales']:
        percent = data.pivot_table(index=index, values=column, aggfunc='sum')\
                                .sort_values(column, ascending=False)\
                                .reset_index()
        percent['per_sales'] = round(percent[column] / percent[column].sum(), 2)
        print(percent.loc[:4])
per_region = count_percent(games_new,'platform')
```

# Шаг 5. Проверка гипотез<a id="step5"></a>


### Гипотеза 1. Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

H0 (нулевая гипотеза): Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

H1 (альтернативная гипотеза): Средние пользовательские рейтинги платформ Xbox One и PC неодинаковые.

Для проверки гипотезы воспользуемся методом scipy.stats.ttest_ind, так как выборки независимы.

```python
#  датафрейм с пользовательскими оценками для платформы xone
xone = games_new.query('platform == "xone" and user_score >= 0')['user_score']
#  датафрейм с пользовательскими оценками для платформы pc
pc = games_new.query('platform == "pc" and user_score >= 0')['user_score']

# гистограмма распределения 
sns.histplot(data=xone, color='#ff0000', alpha=.7, kde=True)
sns.histplot(data=pc, color='#008000', alpha=.5, kde=True)
# установим заголовок для графика
plt.title("Гистограмма распределения")
# подпишем оси координат обозначив единицы измерения
plt.ylabel('Частота')
plt.xlabel('Оценка пользователей')

print('Дисперсия выборки платформы xone = {} '.format(np.var(xone, ddof=1)))
print('Дисперсия выборки платформы pc = {} '.format(np.var(pc, ddof=1)))

stat,p = levene(xone,pc)
print('Тест Левена на равенство дисперсий', p)
```

```python
# критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу
alpha = .05

results = st.ttest_ind(
    xone, 
    pc, 
    equal_var=True)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Можно утверждать с уверенностью 99,99%, что средние пользовательские рейтинги платформ Xbox One и PC неодинаковые. Хотя если взять уровень статистической значимости 0.01, то нулевую гипотезу не получилось бы отвергнуть. 

### Гипотеза 2. Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные

H0 (нулевая гипотеза): Средние пользовательские рейтинги жанров Action и Sports одинаковые.

H1 (альтернативная гипотеза): Средние пользовательские рейтинги жанров Action и Sports неодинаковые.

Для проверки гипотезы воспользуемся методом scipy.stats.ttest_ind, так как выборки независимы.

```python
action = games_new.query('genre == "action" and user_score >= 0')['user_score']
sports = games_new.query('genre == "sports" and user_score >= 0')['user_score']
# гистограмма распределения 
sns.histplot(data=action, color='#ff0000', alpha=.5, kde=True)
sns.histplot(data=sports, color='#008000', alpha=.7, kde=True)
# установим заголовок для графика
plt.title("Гистограмма распределения")
# подпишем оси координат обозначив единицы измерения
plt.ylabel('Частота')
plt.xlabel('Оценка пользователей')

print('Дисперсия выборки жанра action = {} '.format(np.var(action, ddof=1)))
print('Дисперсия выборки жанра sports = {} '.format(np.var(sports, ddof=1)))

stat,p = levene(action,sports)
print('Тест Левена на равенство дисперсий', p)
```

```python
# критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу
alpha = .05

results = st.ttest_ind(
    action, 
    sports, 
    equal_var=False)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Полученное значение p-value говорит о том, что cредние пользовательские рейтинги жанров Action и Sports неодинаковые.
