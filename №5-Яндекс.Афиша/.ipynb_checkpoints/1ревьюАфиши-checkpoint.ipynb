1. –≤ —Ç–∞–±–ª–∏—Ü–∞—Ö visits –∏ orders –Ω–∞–∑–≤–∞–Ω–∏—è —Å—Ç–æ–ª–±—Ü–æ–≤ –ø—Ä–∏–≤–µ—Å—Ç–∏ –∫ –Ω–∏–∂–Ω–µ–º—É —Ä–µ–≥–∏—Å—Ç—Ä—É –∏ –ø–µ—Ä–µ–∏–º–µ–Ω–æ–≤–∞—Ç—å

    * Device –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –ø—Ä–∏–≤–µ—Å—Ç–∏ —Å—Ç–æ–ª–±—Ü–µ—Ü –∫ —Ç–∏–ø—É category
    * End Ts –∏ Start Ts- –∏–∑–º–µ–Ω–∏—Ç—å —Ç–∏–ø –Ω–∞ datetime
    * Source Id - –ø—Ä–∏–≤–µ—Å—Ç–∏ —Ç–∏–ø –∫ int8

    * Buy Ts - –∏–∑–º–µ–Ω–∏—Ç—å –Ω–∞ —Ç–∏–ø datetime
1. –ø–æ–º–µ–Ω—è—Ç—å —Ç–∏–ø –¥–∞–Ω–Ω—ã—Ö –≤ —Ç–∞–±–ª–∏—Ü–µ costs:

# –®–∞–≥ 2. –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö  <a id="step2"></a>

–ü—Ä–∏–≤–µ–¥–µ–º –Ω–∞–∑–≤–∞–Ω–∏—è —Å—Ç–æ–ª–±—Ü–æ–≤ —Ç–∞–±–ª–∏—Ü visits –∏ orders –∫ –Ω–∏–∂–Ω–µ–º—É —Ä–µ–≥–∏—Å—Ç—Ä—É.

```python
# —Ñ—É–Ω–∫—Ü–∏—è –∑–∞–º–µ–Ω—ã –Ω–∞–∑–≤–∞–Ω–∏–π —Å—Ç–æ–ª–±—Ü–æ–≤ –∏ –ø—Ä–∏–≤–µ–¥–µ–Ω–∏—è –∫ –Ω–∏–∂–Ω–µ–º—É —Ä–µ–≥–∏—Å—Ç—Ä—É
def replace_names(data):
    list_of_columns = []
    for column in data.columns:
        column = column.lower()
        column = column.replace(' ', '_')
        list_of_columns.append(column)
    data.columns = list_of_columns
    return data
# –∑–∞–º–µ–Ω–∏–º –Ω–∞–∑–≤–∞–Ω–∏—è —Å—Ç–æ–ª–±—Ü–æ–≤ –≤ —Ç–∞–±–ª–∏—Ü–µ visits
visits = replace_names(visits)
# –≤—ã–≤–æ–¥ –Ω–∞–∑–≤–∞–Ω–∏–π —Å—Ç–æ–ª–±—Ü–æ–≤
visits.columns
```

```python
# –∑–∞–º–µ–Ω–∏–º –Ω–∞–∑–≤–∞–Ω–∏—è —Å—Ç–æ–ª–±—Ü–æ–≤ –≤ —Ç–∞–±–ª–∏—Ü–µ orders
orders = replace_names(orders)
# –≤—ã–≤–æ–¥ –Ω–∞–∑–≤–∞–Ω–∏–π —Å—Ç–æ–ª–±—Ü–æ–≤
orders.columns
```

–¢–µ–ø–µ—Ä—å –Ω–∞–∑–≤–∞–Ω–∏—è —Å—Ç–æ–ª–±—Ü–æ–≤ —Ç–∞–±–ª–∏—Ü visits –∏ orders –ø—Ä–∏–≤–µ–¥–µ–Ω—ã –∫ –Ω–∏–∂–Ω–µ–º—É —Ä–µ–≥–∏—Å—Ç—Ä—É –∏ –ø–µ—Ä–µ–∏–º–µ–Ω–æ–≤–∞–Ω—ã, –ø—Ä–æ–±–µ–ª—ã –≤ –Ω–∞–∑–≤–∞–Ω–∏—è—Ö –∑–∞–º–µ–Ω–µ–Ω—ã –Ω–∞ –Ω–∏–∂–Ω–µ–µ –ø–æ–¥—á–µ—Ä–∫–∏–≤–∞–Ω–∏–µ, –¥–ª—è —É–¥–æ–±—Å—Ç–≤–∞ —Ä–∞–±–æ—Ç—ã.

–î–∞–ª–µ–µ –∏–∑–º–µ–Ω–∏–º —Ç–∏–ø —Å—Ç–æ–ª–±—Ü–æ–≤ —Ç–∞–º –≥–¥–µ —ç—Ç–æ —Ç—Ä–µ–±–æ–≤–∞–ª–æ—Å—å.

```python
# –∏–∑–º–µ–Ω–µ–Ω–∏–µ —Ç–∏–ø–∞ –¥–∞–Ω–Ω—ã—Ö –≤ —Å—Ç–æ–ª–±—Ü–∞—Ö
visits['device'] = visits['device'].astype('category')
print('–ü—Ä–∏–≤–µ–ª–∏ —Å—Ç–æ–ª–±–µ—Ü device –∫ —Ç–∏–ø—É',visits.device.dtypes) 

visits['start_ts'] = pd.to_datetime(visits['start_ts'])
visits['end_ts'] = pd.to_datetime(visits['end_ts'])
print('–ü—Ä–∏–≤–µ–ª–∏ —Å—Ç–æ–ª–±–µ—Ü start_ts –∫ —Ç–∏–ø—É',visits.start_ts.dtypes) 
print('–ü—Ä–∏–≤–µ–ª–∏ —Å—Ç–æ–ª–±–µ—Ü end_ts –∫ —Ç–∏–ø—É',visits.end_ts.dtypes) 

visits['source_id'] = visits['source_id'].astype('int8')
costs['source_id'] = costs['source_id'].astype('int8')
print('–ü—Ä–∏–≤–µ–ª–∏ —Å—Ç–æ–ª–±–µ—Ü source_id –∫ —Ç–∏–ø—É –≤ —Ç–∞–±–ª–∏—Ü–∞—Ö visits –∏ costs',visits.source_id.dtypes) 

orders['buy_ts'] = pd.to_datetime(orders['buy_ts'])
print('–ü—Ä–∏–≤–µ–ª–∏ —Å—Ç–æ–ª–±–µ—Ü buy_ts –∫ —Ç–∏–ø—É',orders.buy_ts.dtypes) 

costs['dt'] = pd.to_datetime(costs['dt'])
print('–ü—Ä–∏–≤–µ–ª–∏ —Å—Ç–æ–ª–±–µ—Ü dt –∫ —Ç–∏–ø—É',costs.dt.dtypes) 
```

### –í—ã–≤–æ–¥

–í —Ö–æ–¥–µ –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∏ –¥–∞–Ω–Ω—ã—Ö –ø—Ä–∏–≤–µ–ª–∏ —Å—Ç–æ–ª–±—Ü—ã –∫ –Ω—É–∂–Ω–æ–º—É —Ç–∏–ø—É, –ø–µ—Ä–µ–∏–º–µ–Ω–æ–≤–∞–ª–∏ –∏ –ø—Ä–∏–≤–µ–ª–∏ –∫ –Ω–∏–∂–Ω–µ–º—É —Ä–µ–≥–∏—Å—Ç—Ä—É –Ω–∞–∑–≤–∞–Ω–∏—è —Å—Ç–æ–ª–±—Ü–æ–≤. –ü–µ—Ä–µ—Ö–æ–¥–∏–º –∫ —à–∞–≥—É 3 –†–∞—Å—Å—á–µ—Ç –º–µ—Ç—Ä–∏–∫ –∏ –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–æ–≤.

## –®–∞–≥ 3.1. –ü—Ä–æ–¥—É–∫—Ç–æ–≤—ã–µ –º–µ—Ç—Ä–∏–∫–∏ <a id="step4"></a>

–†–∞—Å—Å—á–∏—Ç–∞–µ–º DAU, WAU –∏ MAU. –í—ã—á–∏—Å–ª–∏–º —Å—Ä–µ–¥–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è —ç—Ç–∏—Ö –º–µ—Ç—Ä–∏–∫ –∑–∞ –≤–µ—Å—å –ø–µ—Ä–∏–æ–¥.

* DAU (–æ—Ç –∞–Ω–≥–ª. daily active users) ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤ –¥–µ–Ω—å;
* WAU (–æ—Ç –∞–Ω–≥–ª. weekly active users) ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤ –Ω–µ–¥–µ–ª—é;
* MAU (–æ—Ç –∞–Ω–≥–ª. monthly active users) ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤ –º–µ—Å—è—Ü.

–ß—Ç–æ–±—ã –ø–æ—Å—á–∏—Ç–∞—Ç—å –Ω–µ–¥–µ–ª—å–Ω–æ–µ –∏ –º–µ—Å—è—á–Ω–æ–µ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ, –≤—ã–¥–µ–ª–∏–º –≤ –æ—Ç–¥–µ–ª—å–Ω—ã–µ —Å—Ç–æ–ª–±—Ü—ã –≥–æ–¥, –º–µ—Å—è—Ü –∏ –Ω–µ–¥–µ–ª—é, –∞ —Ç–∞–∫–∂–µ –ø–æ–ª–Ω—É—é –¥–∞—Ç—É.

```python
# –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ –≥–æ–¥–∞, –º–µ—Å—è—Ü–∞, –Ω–µ–¥–µ–ª–∏ –∏ –¥–∞—Ç—ã –∏–∑ start_ts —Ç–∞–±–ª–∏—Ü—ã visits
visits['session_year']  = visits['start_ts'].astype('datetime64[Y]')
visits['session_month'] = visits['start_ts'].astype('datetime64[M]')
visits['session_week']  = visits['start_ts'].astype('datetime64[W]')
visits['session_date'] = visits['start_ts'].astype('datetime64[D]')
visits.sample(5) 
```

–°–≥—Ä—É–ø–ø–∏—Ä—É–µ–º –¥–∞–Ω–Ω—ã–µ –ø–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã–º –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è–º, –Ω–∞–π–¥—ë–º —Å—Ä–µ–¥–Ω–µ–µ –∏ –ø–æ—Å—á–∏—Ç–∞–µ–º –º–µ—Ç—Ä–∏–∫–∏.

```python
dau_total = visits.groupby('session_date').agg({'uid': 'nunique'})
wau_total = visits.groupby(['session_year', 'session_week']).agg({'uid': 'nunique'})
mau_total = visits.groupby(['session_year','session_month']).agg({'uid':'nunique'})
print('–í –¥–µ–Ω—å —Å–µ—Ä–≤–∏—Å–æ–º –≤ —Å—Ä–µ–¥–Ω–µ–º –ø–æ–ª—å–∑—É—é—Ç—Å—è {} –ª—é–¥–µ–π'.format(int(dau_total.mean())))
print('–í –Ω–µ–¥–µ–ª—é —Å–µ—Ä–≤–∏—Å–æ–º –≤ —Å—Ä–µ–¥–Ω–µ–º –ø–æ–ª—å–∑—É—é—Ç—Å—è {} –ª—é–¥–µ–π'.format(int(wau_total.mean())))
print('–í –º–µ—Å—è—Ü —Å–µ—Ä–≤–∏—Å–æ–º –≤ —Å—Ä–µ–¥–Ω–µ–º –ø–æ–ª—å–∑—É—é—Ç—Å—è {} –ª—é–¥–µ–π'.format(int(mau_total.mean())))
```

</div>

```python
dau_total.plot.line(figsize = (16, 6), title = '–°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ—Å–µ—â–µ–Ω–∏–π –Ø–Ω–¥–µ–∫—Å.–ê—Ñ–∏—à–∞ –≤ –¥–µ–Ω—å')
plt.xlabel("–î–∞—Ç—ã")
plt.ylabel("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ—Å–µ—â–µ–Ω–∏–π")
plt.show()
```

</div>

<div class="alert alert-info"> <b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Å—Ç—É–¥–µ–Ω—Ç–∞:</b> –î—É–º–∞—é, —á—Ç–æ –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å –Ω–æ—Ä–º–∞–ª—å–Ω—ã–π. –°—É–±—ä–µ–∫—Ç–∏–≤–Ω–æ, –º–Ω–µ –∫–∞–∂–µ—Ç—Å—è, —á—Ç–æ –≤ –†–æ—Å—Å–∏–∏ –¥–∞–ª–µ–∫–æ –Ω–µ –≤—Å–µ –º–æ–≥—É—Ç –ø–æ–∑–≤–æ–ª–∏—Ç—å —Å–µ–±–µ —Ö–æ–¥–∏—Ç—å –Ω–∞ —Ä–∞–∑–≤–ª–µ–∫–∞—Ç–µ–ª—å–Ω—ã–µ –º–µ—Ä–æ–ø—Ä–∏—è—Ç–∏—è —á–∞—â–µ</div>

```python
# –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ —Å—Ç–æ–ª–±—Ü–∞ —Å–æ –¥–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å—é –ø–æ—Å–µ—â–µ–Ω–∏—è —Å–∞–π—Ç–∞ –≤ —Å–µ–∫—É–Ω–¥–∞—Ö
visits['session_duration_sec'] = (visits['end_ts'] - visits['start_ts']).dt.seconds
```

```python
# –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è 
sns.histplot(data=visits, x='session_duration_sec', color='#ff0000', alpha=.7, bins = 100, kde=True)
# —É—Å—Ç–∞–Ω–æ–≤–∏–º –∑–∞–≥–æ–ª–æ–≤–æ–∫ –¥–ª—è –≥—Ä–∞—Ñ–∏–∫–∞
plt.title("–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –¥–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ —Å–µ—Å—Å–∏–π")
# –ø–æ–¥–ø–∏—à–µ–º –æ—Å–∏ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç –æ–±–æ–∑–Ω–∞—á–∏–≤ –µ–¥–∏–Ω–∏—Ü—ã –∏–∑–º–µ—Ä–µ–Ω–∏—è
plt.ylabel('–ß–∞—Å—Ç–æ—Ç–∞')
plt.xlabel('–î–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å —Å–µ—Å—Å–∏–∏, —Å–µ–∫.')
plt.show()
```

–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –Ω–µ –ø–æ—Ö–æ–∂–µ –Ω–∞ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–µ, –ø–æ—ç—Ç–æ–º—É —Ä–∞—Å—Å—á–∏—Ç–∞–µ–º –º–æ–¥—É (–∑–Ω–∞—á–µ–Ω–∏–µ –≤ –≤—ã–±–æ—Ä–∫–µ, –∫–æ—Ç–æ—Ä–æ–µ –≤—Å—Ç—Ä–µ—á–∞–µ—Ç—Å—è —á–∞—â–µ –≤—Å–µ–≥–æ):

C—Ä–µ–¥–Ω—è—è –ø—Ä–æ–¥–æ–ª–∂–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å —Å–µ—Å—Å–∏–∏ 60 —Å–µ–∫—É–Ω–¥, –Ω–æ –ø–æ–∫–∞ –º—ã –Ω–µ –º–æ–∂–µ–º –æ—Ü–µ–Ω–∏—Ç—å –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –ª–∏ —ç—Ç–æ –¥–ª—è –¥–æ—Å—Ç–∏–∂–µ–Ω–∏—è –∫–æ–Ω–µ—á–Ω–æ–π —Ü–µ–ª–∏ –∫–ª–∏–µ–Ω—Ç–∞(–ø–æ–∫—É–ø–∫–∏ –±–∏–ª–µ—Ç–∞). 

<div class="alert alert-success">
<b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞:</b> –°–æ–≥–ª–∞—Å–µ–Ω —Å —Ç–æ–±–æ–π. –ö—Å—Ç–∞—Ç–∏, –æ—á–µ–Ω—å –≥—Ä–∞–º–æ—Ç–Ω—ã–π –≤—ã–±–æ—Ä –º–µ—Ç—Ä–∏–∫–∏, –º–æ–¥–∞ –∑–¥–µ—Å—å –æ–ø—Ç–∏–º–∞–ª—å–Ω–∞.
</div>

–†–∞—Å—Å—á–∏—Ç–∞–µ–º Retention Rate(–∞–Ω–≥–ª. ¬´–∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç —É–¥–µ—Ä–∂–∞–Ω–∏—è¬ª), –ø—Ä–∏–º–µ–Ω—è—è –∫–æ–≥–æ—Ä—Ç–Ω—ã–π –∞–Ω–∞–ª–∏–∑. –û–ø—Ä–µ–¥–µ–ª–∏–º —Å–æ–±—ã—Ç–∏–µ –∏ –ø–µ—Ä–∏–æ–¥ ‚Äî –Ω–∞ –∏—Ö –æ—Å–Ω–æ–≤–µ —Å—Ñ–æ—Ä–º–∏—Ä—É–µ–º –∫–æ–≥–æ—Ä—Ç—É. –í–æ–∑—å–º—ë–º –¥–∞—Ç—É, –∫–æ–≥–¥–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å –≤–ø–µ—Ä–≤—ã–µ –ø—Ä–æ—è–≤–∏–ª –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç—å –≤ —Å–µ—Ä–≤–∏—Å–µ –Ø–Ω–¥–µ–∫—Å.–ê—Ñ–∏—à–∞.

```python
# —Ä–∞—Å—á–µ—Ç –¥–∞—Ç—ã, –∫–æ–≥–¥–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å –ø—Ä–æ—è–≤–∏–ª –ø–µ—Ä–≤—É—é –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç—å
first_activity = visits.groupby('uid').agg({'start_ts':'min'})
# # –ø–µ—Ä–µ–∏–º–µ–Ω–æ–≤–∞–Ω–∏–µ —Å—Ç–æ–ª–±—Ü–∞
first_activity.columns = ['first_activity']
visits = visits.join(first_activity, on='uid')
```

–¢–µ–ø–µ—Ä—å –¥–ª—è –∫–∞–∂–¥–æ–π —Å—Ç—Ä–æ–∫–∏ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ –º–æ–∂–Ω–æ —Ä–∞—Å—Å—á–∏—Ç–∞—Ç—å lifetime –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –≤ —Ä–∞–º–∫–∞—Ö –∫–æ–≥–æ—Ä—Ç—ã. 

```python
# –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ —Å—Ç–æ–ª–±—Ü–∞ first_activity_month, –º–µ—Å—è—Ü –ø–µ—Ä–≤–æ–≥–æ –ø–æ—Å–µ—â–µ–Ω–∏—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è
visits['first_activity_month'] = visits['first_activity'].astype('datetime64[M]')
# –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ —Å—Ç–æ–ª–±—Ü–∞ start_ts_month, –º–µ—Å—è—Ü –Ω–∞—á–∞–ª–∞ —Å–µ—Å—Å–∏–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è
visits['start_ts_month'] = visits['start_ts'].astype('datetime64[M]')
# —Ä–∞—Å—á–µ—Ç lifetime
visits['lifetime'] = ((visits['start_ts_month'] - visits['first_activity_month']) / np.timedelta64(1, 'M')).round().astype('int')
```

–°–≥—Ä—É–ø–ø–∏—Ä—É–µ–º –¥–∞–Ω–Ω—ã–µ –ø–æ –∫–æ–≥–æ—Ä—Ç–µ –∏ lifetime. –ü–æ—Å—á–∏—Ç–∞–µ–º –¥–ª—è –∫–∞–∂–¥–æ–π –∫–æ–≥–æ—Ä—Ç—ã –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∞–∫—Ç–∏–≤–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –Ω–∞ –æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω—ã–π ¬´–º–µ—Å—è—Ü –∂–∏–∑–Ω–∏¬ª:

```python
cohorts = visits.groupby(['first_activity_month', 'lifetime']).agg({'uid':'nunique'}).reset_index()
initial_users_count = cohorts[cohorts['lifetime'] == 0][['first_activity_month', 'uid']]
initial_users_count.columns = ['first_activity_month', 'cohort_users']
initial_users_count = initial_users_count.set_index('first_activity_month')
```

–û–±—ä–µ–¥–∏–Ω–∏–º –¥–∞–Ω–Ω—ã–µ –ø–æ –∫–æ–≥–æ—Ä—Ç–∞–º —Å –∏—Å—Ö–æ–¥–Ω—ã–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤ –∫–æ–≥–æ—Ä—Ç–µ:

```python
cohorts = cohorts.merge(initial_users_count,on='first_activity_month') 
```

–ù–∞–∫–æ–Ω–µ—Ü, —Ä–∞—Å—Å—á–∏—Ç–∞–µ–º Retention Rate. –†–∞–∑–¥–µ–ª–∏–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∞–∫—Ç–∏–≤–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤ –∫–∞–∂–¥—É—é –∏–∑ –º–µ—Å—è—Ü–æ–≤ –Ω–∞ –∏—Å—Ö–æ–¥–Ω–æ–µ —á–∏—Å–ª–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤ –∫–æ–≥–æ—Ä—Ç–µ:

```python
cohorts['retention'] = cohorts['uid']/cohorts['cohort_users'] 
```

–ü–æ—Å—Ç—Ä–æ–∏–º —Å–≤–æ–¥–Ω—É—é —Ç–∞–±–ª–∏—Ü—É –∏ —Å–æ–∑–¥–∞–¥–∏–º —Ç–µ–ø–ª–æ–≤—É—é –∫–∞—Ä—Ç—É:

```python
retention_pivot = cohorts.pivot_table(index='first_activity_month',\
                                      columns='lifetime',\
                                      values='retention',\
                                      aggfunc='sum') 
```

–ü–æ—Å—Ç—Ä–æ–∏–º —Ç–µ–ø–ª–æ–≤—É—é –∫–∞—Ä—Ç—É:

```python
sns.set(style='whitegrid')
# —Ä–∞–∑–º–µ—Ä —Ç–µ–ø–ª–æ–≤–æ–π –∫–∞—Ä—Ç—ã
plt.figure(figsize=(13, 9))
# –∑–∞–≥–æ–ª–æ–≤–æ–∫ —Ç–µ–ø–ª–æ–≤–æ–π –∫–∞—Ä—Ç—ã
plt.title('Cohorts: User Retention')
# –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ —Ç–µ–ø–ª–æ–≤–æ–π –∫–∞—Ä—Ç—ã
sns.heatmap(retention_pivot, annot=True, fmt='0.1%', linewidths=1, linecolor='black', vmin = 0.01, vmax = 0.1);
# –ø–æ–¥–ø–∏—à–µ–º –æ—Å–∏ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç –æ–±–æ–∑–Ω–∞—á–∏–≤ –µ–¥–∏–Ω–∏—Ü—ã –∏–∑–º–µ—Ä–µ–Ω–∏—è
plt.ylabel('–ü–µ—Ä–≤—ã–π –º–µ—Å—è—Ü –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏')
plt.xlabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –º–µ—Å—è—Ü–µ–≤ —Å –ø–µ—Ä–≤–æ–π –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏')
```

–ë–æ–ª—å—à–µ –≤—Å–µ–≥–æ —Å–≤–æ–∏—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –ø–æ—Ç–µ—Ä—è–ª–∞ –∫–æ–≥–æ—Ä—Ç–∞ –¥–µ–∫–∞–±—Ä—è, –≤ –∫–æ—Ç–æ—Ä–æ–π –∫ –º–∞—é –æ—Å—Ç–∞–ª–æ—Å—å 1.9%.

## –®–∞–≥ 3.2. –ú–µ—Ç—Ä–∏–∫–∏ —ç–ª–µ–∫—Ç—Ä–æ–Ω–Ω–æ–π –∫–æ–º–º–µ—Ä—Ü–∏–∏ <a id="step5"></a>

–ò—Å—Å–ª–µ–¥—É–µ–º, —Å–∫–æ–ª—å–∫–æ –≤—Ä–µ–º–µ–Ω–∏ –≤ —Å—Ä–µ–¥–Ω–µ–º –ø—Ä–æ—Ö–æ–¥–∏—Ç —Å –º–æ–º–µ–Ω—Ç–∞ –ø–µ—Ä–≤–æ–≥–æ –ø–æ—Å–µ—â–µ–Ω–∏—è —Å–∞–π—Ç–∞ –¥–æ —Å–æ–≤–µ—Ä—à–µ–Ω–∏—è –ø–æ–∫—É–ø–∫–∏. –î–ª—è —ç—Ç–æ–≥–æ –Ω–∞–π–¥–µ–º –≤—Ä–µ–º—è –ø–µ—Ä–≤–æ–π –ø–æ–∫—É–ø–∫–∏ –≤ —Ç–∞–±–ª–∏—Ü–µ orders –∏ –¥–æ–±–∞–≤–∏–º –¥–∞—Ç—É –∏ –≤—Ä–µ–º—è –ø–µ—Ä–≤–æ–π —Å–µ—Å—Å–∏–∏ –≤ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –ø–æ –ø–µ—Ä–≤—ã–º –∑–∞–∫–∞–∑–∞–º. –ü–æ—Å–ª–µ –≤—ã—á–∏—Å–ª–∏–º —Ä–∞–∑–Ω–∏—Ü—É –∏ —É–∑–Ω–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—Ä–µ–º–µ–Ω–∏, –∫–æ—Ç–æ—Ä–æ–µ –ø—Ä–æ—à–ª–æ –¥–æ –º–æ–º–µ–Ω—Ç–∞ —Å–æ–≤–µ—Ä—à–µ–Ω–∏—è –ø–µ—Ä–≤–æ–π –ø–æ–∫—É–ø–∫–∏ –∫–ª–∏–µ–Ω—Ç–æ–º.

```python
# —Ä–∞—Å—á–µ—Ç –¥–∞—Ç—ã –∏ –≤—Ä–µ–º–µ–Ω–∏ –ø–µ—Ä–≤–æ–π –ø–æ–∫—É–ø–∫–∏ –∫–ª–∏–µ–Ω—Ç–∞
first_orders = orders.groupby(['uid']).agg({'buy_ts':'min'})
# –ø–µ—Ä–µ–∏–º–µ–Ω–æ–≤–∞–Ω–∏–µ —Å—Ç–æ–ª–±—Ü–∞
first_orders.columns = ['first_orders']
# –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ —Ç–∞–±–ª–∏—Ü orders –∏ first_orders
orders_ = orders.join(first_orders, on='uid')
```

<div class="alert alert-warning">
<b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞: </b> –°–ø–æ—Ä–Ω–æ —Ä–µ—à–µ–Ω–∏–µ, —Ç–∞–∫ –∫–∞–∫ –≤ —Å–ª—É—á–∞–µ –æ—à–∏–±–∫–∏ —Ç—ã —É–±–µ—Ä–µ—à—å –µ–µ –ª–æ–≥–∏. –ë—É–¥–µ—Ç —Å–ª–æ–∂–Ω–µ–µ –ø–æ–Ω—è—Ç—å —á—Ç–æ —Ç–∞–º —Å–ª–æ–º–∞–ª–æ—Å—å.

<b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Å—Ç—É–¥–µ–Ω—Ç–∞: </b> –£–±—Ä–∞–ª–∞ –∫–æ–Ω—Å—Ç—Ä—É–∫—Ü–∏—é try|except</code>
</div>

–î–æ–±–∞–≤–∏–º –≤ —Ç–∞–±–ª–∏—Ü—É orders —Å—Ç–æ–ª–±–µ—Ü —Å –¥–∞—Ç–æ–π –ø–µ—Ä–≤–æ–π –∞–∫—Ç–∏–≤–Ω–æ–π —Å–µ—Å—Å–∏–∏, –∫–æ—Ç–æ—Ä—É—é –º—ã —É–∂–µ —Å—á–∏—Ç–∞–ª–∏ –≤—ã—à–µ, –¥–∞–Ω–Ω—ã–µ —Ö—Ä–∞–Ω—è—Ç—Å—è –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ first_activity, –ø—É—Ç–µ–º –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏—è —Ç–∞–±–ª–∏—Ü orders –∏ first_activity –ø–æ –∫–ª—é—á—É uid.

```python
# –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ —Ç–∞–±–ª–∏—Ü orders –∏ first_activity
orders_ = orders_.join(first_activity, on='uid')
```

–ù–∞–π–¥–µ–º –≤—Ä–µ–º—è, –ø—Ä–æ—à–µ–¥—â–µ–µ —Å –º–æ–º–µ–Ω—Ç–∞ –ø–µ—Ä–≤–æ–π —Å–µ—Å—Å–∏–∏ –¥–æ –ø–µ—Ä–≤–æ–π –ø–æ–∫—É–ø–∫–∏.

```python
orders_['time_to_first_purchase'] = (orders_['first_orders'] - orders_['first_activity']) / np.timedelta64(1, 'm')
print('–ú–µ–¥–∏–∞–Ω–∞ –≤—Ä–µ–º–µ–Ω–∏ —Å –ø–µ—Ä–≤–æ–π —Å–µ—Å—Å–∏–∏ –¥–æ –ø–µ—Ä–≤–æ–π –ø–æ–∫—É–ø–∫–∏: {} –º–∏–Ω.'.format(int(orders_['time_to_first_purchase'].median())))
```

–ò–∑ –≥—Ä–∞—Ñ–∏–∫–∞ –Ω–∞–±–ª—é–¥–∞–µ–º, —á—Ç–æ –±–æ–ª—å—à–∏–Ω—Å—Ç–≤–æ –∫–ª–∏–µ–Ω—Ç–æ–≤ —Å–æ–≤–µ—Ä—à–∞—é—Ç –ø–µ—Ä–≤—É—é –ø–æ–∫—É–ø–∫—É –¥–æ 10 –º–∏–Ω—É—Ç. –ü–∏–∫ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –≤ —Ä–∞–æ–π–Ω–µ –º–µ–Ω—å—à–µ 1 –º–∏–Ω—É—Ç—ã. –ú–µ–¥–∏–∞–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –≥–æ–≤–æ—Ä–∏—Ç –Ω–∞–º –æ —Ç–æ–º, —á—Ç–æ –≤—Ä–µ–º—è –æ—Ç –ø–µ—Ä–≤–æ–≥–æ –ø–æ—Å–µ—â–µ–Ω–∏—è –¥–æ —Å–æ–≤–µ—Ä—à–µ–Ω–∏—è –ø–æ–∫—É–ø–∫–∏ –æ–∫–æ–ª–æ 22 –º–∏–Ω—É—Ç. –í—ã—à–µ –º—ã –ø–æ–ª—É—á–∞–ª–∏, —á—Ç–æ —Å—Ä–µ–¥–Ω—è—è –ø—Ä–æ–¥–æ–ª–∂–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å —Å–µ—Å—Å–∏–∏ 60 —Å–µ–∫—É–Ω–¥, –∑–∞ —ç—Ç–æ –≤—Ä–µ–º—è —É—Å–ø–µ–≤–∞—é—Ç —Å–¥–µ–ª–∞—Ç—å –æ–∫–æ–ª–æ 4000 –ø–æ–∫—É–ø–æ–∫ –Ω–∞ —Å–∞–π—Ç–µ –Ø–Ω–¥–µ–∫—Å.–ê—Ñ–∏—à–∞.

</div>

```python
# –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ —Å—Ç–æ–ª–±—Ü–∞ —Å –¥–Ω–µ–º —Å–æ–≤–µ—Ä—à–µ–Ω–∏—è –ø–æ–∫—É–ø–∫–∏
orders_['buy_date'] = orders_['buy_ts'].dt.date
# –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ —Å—Ç–æ–ª–±—Ü–∞ —Å –Ω–µ–¥–µ–ª–µ–π —Å–æ–≤–µ—Ä—à–µ–Ω–∏—è –ø–æ–∫—É–ø–∫–∏
orders_['buy_week'] = orders_['buy_ts'].dt.week
# –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ —Å—Ç–æ–ª–±—Ü–∞ —Å –º–µ—Å—è—Ü–æ–º —Å–æ–≤–µ—Ä—à–µ–Ω–∏—è –ø–æ–∫—É–ø–∫–∏
orders_['buy_month'] = orders_['buy_ts'].dt.month
# –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ —Å—Ç–æ–ª–±—Ü–∞ —Å –≥–æ–¥–æ–º —Å–æ–≤–µ—Ä—à–µ–Ω–∏—è –ø–æ–∫—É–ø–∫–∏
orders_['buy_year'] = orders_['buy_ts'].dt.year

day_purchases_per_user = orders_.groupby(['buy_date', 'uid']).agg({'buy_ts':'count'}).mean()[0]
week_purchases_per_user = orders_.groupby(['buy_year', 'buy_week', 'uid']).agg({'buy_ts':'count'}).mean()[0]
month_purchases_per_user = orders_.groupby(['buy_month', 'uid']).agg({'buy_ts':'count'}).mean()[0]
year_purchases_per_user = orders_.groupby(['buy_year', 'uid']).agg({'buy_ts':'count'}).mean()[0]
print('–í —Å—Ä–µ–¥–Ω–µ–º –Ω–∞ –æ–¥–Ω–æ–≥–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –ø–æ–∫—É–ø–æ–∫ –≤ –¥–µ–Ω—å {:.2f}'.format(day_purchases_per_user))
print('–í —Å—Ä–µ–¥–Ω–µ–º –Ω–∞ –æ–¥–Ω–æ–≥–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –ø–æ–∫—É–ø–æ–∫ –≤ –º–µ—Å—è—Ü {:.2f}'.format(month_purchases_per_user))
print('–í —Å—Ä–µ–¥–Ω–µ–º –Ω–∞ –æ–¥–Ω–æ–≥–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –ø–æ–∫—É–ø–æ–∫ –≤ –Ω–µ–¥–µ–ª—é {:.2f}'.format(week_purchases_per_user))
print('–í —Å—Ä–µ–¥–Ω–µ–º –Ω–∞ –æ–¥–Ω–æ–≥–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –ø–æ–∫—É–ø–æ–∫ –≤ –≥–æ–¥ {:.2f}'.format(year_purchases_per_user))
```

–£–∑–Ω–∞–µ–º, –∫–∞–∫ –∏–∑–º–µ–Ω—è–µ—Ç—Å—è –ø–æ–≤–µ–¥–µ–Ω–∏–µ –∫–∞–∂–¥–æ–π –∫–æ–≥–æ—Ä—Ç—ã –Ω–∞ –ø—Ä–æ—Ç—è–∂–µ–Ω–∏–∏ –≤—Ä–µ–º–µ–Ω–∏. –ù–∞–ø—Ä–∏–º–µ—Ä, –æ–ø—Ä–µ–¥–µ–ª–∏–º, —Å–Ω–∏–∂–∞–µ—Ç—Å—è –ª–∏ —Å—Ä–µ–¥–Ω–∏–π —á–µ–∫ –∫–ª–∏–µ–Ω—Ç–æ–≤ –∏ –æ–¥–∏–Ω–∞–∫–æ–≤—ã–π –ª–∏ –¥–æ—Ö–æ–¥ –ø–æ–ª—É—á–∞–µ—Ç –∫–æ–º–ø–∞–Ω–∏—è –æ—Ç –∫–ª–∏–µ–Ω—Ç–æ–≤, –ø—Ä–∏–≤–ª–µ—á—ë–Ω–Ω—ã—Ö –≤ —Ä–∞–∑–Ω—ã–µ –ø–µ—Ä–∏–æ–¥—ã.

–î–ª—è –Ω–∞—á–∞–ª–∞ —Å—Ñ–æ—Ä–º–∏—Ä—É–µ–º –∫–æ–≥–æ—Ä—Ç—ã –ø–æ –º–µ—Å—è—Ü–∞–º –ø–æ–∫—É–ø–∫–∏.

```python
# –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ —Å—Ç–æ–ª–±—Ü–∞ first_activity_month, –º–µ—Å—è—Ü –ø–µ—Ä–≤–æ–≥–æ –ø–æ—Å–µ—â–µ–Ω–∏—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è
orders_['first_activity_month'] = orders_['first_activity'].astype('datetime64[M]')
# –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ —Å—Ç–æ–ª–±—Ü–∞ start_ts_month, –º–µ—Å—è—Ü –Ω–∞—á–∞–ª–∞ —Å–µ—Å—Å–∏–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è
orders_['buy_date'] = orders_['buy_date'].astype('datetime64[M]')
# —Ä–∞—Å—á–µ—Ç lifetime
orders_['lifetime'] = ((orders_['buy_date'] - orders_['first_activity_month']) / \
                      np.timedelta64(1, 'M')).round().astype('int')
cohorts = orders_.groupby(['first_activity_month', 'lifetime'])\
                .agg({'revenue':'sum', 'uid':'nunique'})\
                .reset_index()
# –ø–µ—Ä–µ–∏–º–µ–Ω–æ–≤–∞–Ω–∏–µ —Å—Ç–æ–ª–±—Ü–æ–≤
cohorts.columns = ['first_activity_month', 'lifetime', 'revenue_total', 'n_users']
# –≤—ã—Ä—É—á–∫–∞ —Å –æ–¥–Ω–æ–≥–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è
cohorts['revenue_per_user'] = cohorts['revenue_total'] / cohorts['n_users']
# –∫–æ–≥–æ—Ä—Ç—ã –ø–æ —Å—É–º–º–µ –ø–æ–∫—É–ø–æ–∫ –Ω–∞ –æ–¥–Ω–æ–≥–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è
cohorts_revenue_per_user = cohorts.pivot_table(index = 'first_activity_month',\
                                               columns = 'lifetime', \
                                               values = 'revenue_per_user', \
                                               aggfunc = 'mean')
# —Ä–∞—Å—á–µ—Ç —Å—Ä–µ–¥–Ω–µ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è –ø–æ –≤—Å–µ–º –º–µ—Å—è—Ü–∞–º –∂–∏–∑–Ω–∏ –∫–æ–≥–æ—Ä—Ç—ã
cohorts_revenue_per_user['mean'] = cohorts_revenue_per_user.mean(axis=1)
```

–ü–æ—Å—Ç—Ä–æ–∏–º —Ç–µ–ø–ª–æ–≤—É—é –∫–∞—Ä—Ç—É:

```python
sns.set(style='whitegrid')
# —Ä–∞–∑–º–µ—Ä —Ç–µ–ø–ª–æ–≤–æ–π –∫–∞—Ä—Ç—ã
plt.figure(figsize=(13, 9))
# –∑–∞–≥–æ–ª–æ–≤–æ–∫ —Ç–µ–ø–ª–æ–≤–æ–π –∫–∞—Ä—Ç—ã
plt.title('–ö–æ–≥–æ—Ä—Ç—ã: —Å—É–º–º–∞ –ø–æ–∫—É–ø–æ–∫ –Ω–∞ –æ–¥–Ω–æ–≥–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è')
# –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ —Ç–µ–ø–ª–æ–≤–æ–π –∫–∞—Ä—Ç—ã
sns.heatmap(cohorts_revenue_per_user, annot=True, fmt=".1f", linewidths=.5, cmap = 'RdYlGn');
# –ø–æ–¥–ø–∏—à–µ–º –æ—Å–∏ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç –æ–±–æ–∑–Ω–∞—á–∏–≤ –µ–¥–∏–Ω–∏—Ü—ã –∏–∑–º–µ—Ä–µ–Ω–∏—è
plt.ylabel('–ü–µ—Ä–≤—ã–π –º–µ—Å—è—Ü –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏')
plt.xlabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –º–µ—Å—è—Ü–µ–≤ —Å –ø–µ—Ä–≤–æ–π –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏')
plt.show()
```

–° —Ç–µ—á–µ–Ω–∏–µ–º –≤—Ä–µ–º–µ–Ω–∏ –ø—Ä–∞–∫—Ç–∏—á–µ—Å–∫–∏ –≤—Å–µ –∫–æ–≥–æ—Ä—Ç—ã –¥–µ–º–æ–Ω—Å—Ç—Ä–∏—Ä—É—é—Ç —Ä–æ—Å—Ç —Å—Ä–µ–¥–Ω–µ–≥–æ —á–µ–∫–∞ –ø–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è–º. 

–í—ã—Å–æ–∫–∏–π —Ä–æ—Å—Ç —Å—Ä–µ–¥–Ω–µ–≥–æ —á–µ–∫–∞ –¥–µ–º–æ–Ω—Å—Ç—Ä–∏—Ä—É—é—Ç –∫–æ–≥–æ—Ä—Ç—ã –∏—é–Ω—è, –¥–µ–∫–∞–±—Ä—è –∏ —Å–µ–Ω—Ç—è–±—Ä—è, —Å—Ä–µ–¥–Ω–∏–π —á–µ–∫ –±–æ–ª—å—à–µ 14. –ê –≤ –æ—Å—Ç–∞–ª—å–Ω—ã—Ö –∫–æ–≥–æ—Ä—Ç–∞—Ö, —Å—Ä–µ–¥–Ω–∏–π —á–µ–∫ –æ–∫–æ–ª–æ 4-8. –ó–∞–º–µ—á–∞—Ç–µ–ª—å–Ω–æ, —á—Ç–æ –≤ –∫–æ–≥–æ—Ä—Ç–µ —Å–µ–Ω—Ç—è–±—Ä—è –µ—Å—Ç—å —Ä–µ–∑–∫–∏–π —Å–∫–∞—á–æ–∫ –¥–æ 58.6, –æ—á–µ–≤–∏–¥–Ω–æ –±—ã–ª–∞ —Å–æ–≤–µ—Ä—à–µ–Ω–∞ –∫–∞–∫–∞—è-—Ç–æ –∫—Ä—É–ø–Ω–∞—è –ø–æ–∫—É–ø–∫–∞ –∏–ª–∏ –∏–º–µ—é—Ç—Å—è –∞–Ω–æ–º–∞–ª—å–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –≤ –∏—Å—Ö–æ–¥–Ω–æ–º –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ, –∑–∞ —Å—á–µ—Ç —ç—Ç–æ–≥–æ —Å–µ–Ω—Ç—è–±—Ä—å—Å–∫–∞—è –∫–æ–≥–æ—Ä—Ç–∞ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –≤ –ª–∏–¥–µ—Ä–∞—Ö, –Ω–µ—Å–º–æ—Ç—Ä—è –Ω–∞ —Ç–æ, —á—Ç–æ –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç –ø–∞–¥–µ–Ω–∏–µ –∫ 8–º—É –º–µ—Å—è—Ü—É –æ—Ç 1–≥–æ –º–µ—Å—è—Ü–∞ –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏ (–±—ã–ª–æ 5.5 —Å—Ç–∞–ª–æ 4.5). –¢–∞–∫ –∂–µ –ø–∏–∫–∏ –ø—Ä–æ–¥–∞–∂ –Ω–∞–±–ª—é–¥–∞–µ–º –≤ —Å–µ–Ω—Ç—è–±—Ä–µ –∏ –¥–µ–∫–∞–±—Ä–µ, –Ω–æ –≤ –¥–≤–∞ —Ä–∞–∑–∞ –Ω–∏–∂–µ, –æ–∫–æ–ª–æ 20.

```python
print('–°—Ä–µ–¥–Ω–∏–π —á–µ–∫ —Ä–∞–≤–µ–Ω {:.2f} —É.–µ.'.format(orders_['revenue'].mean()))
```

–¢–µ–ø–µ—Ä—å –ø–æ—Å—á–∏—Ç–∞–µ–º —ç–∫–æ–Ω–æ–º–∏–∫—É –æ–¥–Ω–æ–≥–æ –ø–æ–∫—É–ø–∞—Ç–µ–ª—è.

LTV (–æ—Ç –∞–Ω–≥–ª. lifetime value) ‚Äî ¬´–ø–æ–∂–∏–∑–Ω–µ–Ω–Ω–∞—è¬ª —Ü–µ–Ω–Ω–æ—Å—Ç—å –∫–ª–∏–µ–Ω—Ç–∞. –≠—Ç–æ ‚Äî –∏—Ç–æ–≥–æ–≤–∞—è —Å—É–º–º–∞ –¥–µ–Ω–µ–≥, –∫–æ—Ç–æ—Ä—É—é –æ–¥–∏–Ω –∫–ª–∏–µ–Ω—Ç –ø—Ä–∏–Ω–æ—Å–∏—Ç –≤ —Å—Ä–µ–¥–Ω–µ–º —Å–æ –≤—Å–µ—Ö —Å–≤–æ–∏—Ö –ø–æ–∫—É–ø–æ–∫. –ü—Ä–∏ —Ä–∞—Å—á—ë—Ç–µ LTV –æ–±—ã—á–Ω–æ –æ–±—Ä–∞—â–∞—é—Ç—Å—è –∫ –≤–∞–ª–æ–≤–æ–π –ø—Ä–∏–±—ã–ª–∏ —Å –∫–ª–∏–µ–Ω—Ç–∞ –∑–∞ –ø–µ—Ä–∏–æ–¥.

–ú–∞—Ä–∂–∏–Ω–∞–ª—å–Ω–æ—Å—Ç—å —Å–µ—Ä–≤–∏—Å–∞ ‚Äî 100%.

–ü–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –¥–∞—Ç –∑–∞–∫–∞–∑–æ–≤:

```python
orders_['first_orders'].describe() 
```

–ü–µ—Ä–≤–∞—è –ø–æ–∫—É–ø–∫–∞ —Å–æ—Å—Ç–æ—è–ª–∞—Å—å 1 –∏—é–Ω—è 2017 –≥–æ–¥–∞. –ü–æ—Å–ª–µ–¥–Ω—è—è ‚Äî —Å–ø—É—Å—Ç—è —Ä–æ–≤–Ω–æ –≥–æ–¥ 1 –∏—é–Ω—è 2018 –≥–æ–¥–∞.

```python
costs['dt'].describe() 
```

–î–∏–∞–ø–∞–∑–æ–Ω –¥–∞—Ç —Ä–∞—Å—Ö–æ–¥–æ–≤ —Å–æ–≤–ø–∞–¥–∞–µ—Ç —Å –¥–∞—Ç–∞–º–∏ –∑–∞–∫–∞–∑–æ–≤.

LTV —É–¥–æ–±–Ω–æ —Å—á–∏—Ç–∞—Ç—å –ø–æ –∫–æ–≥–æ—Ä—Ç–∞–º. –î–ª—è —ç—Ç–æ–≥–æ –Ω—É–∂–Ω–æ:
* –Ω–∞–π—Ç–∏ –¥–∞—Ç—É –ø–µ—Ä–≤–æ–π –ø–æ–∫—É–ø–∫–∏ –∫–∞–∂–¥–æ–≥–æ –ø–æ–∫—É–ø–∞—Ç–µ–ª—è;
* –≤—ã—á–∏—Å–ª–∏—Ç—å —á–∏—Å–ª–æ –Ω–æ–≤—ã—Ö –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π –∑–∞ –∫–∞–∂–¥—É—é –¥–∞—Ç—É;
* –¥–æ–±–∞–≤–∏—Ç—å –¥–∞—Ç—É –ø–µ—Ä–≤–æ–π –ø–æ–∫—É–ø–∫–∏ –≤ —Ç–∞–±–ª–∏—Ü—É —Å –∑–∞–∫–∞–∑–∞–º–∏;
* —Å–≥—Ä—É–ø–ø–∏—Ä–æ–≤–∞—Ç—å —Ç–∞–±–ª–∏—Ü—É –∏ –ø–æ—Å—á–∏—Ç–∞—Ç—å –≤—ã—Ä—É—á–∫—É.

–í—ã–¥–µ–ª–∏–º –º–µ—Å—è—Ü—ã –∏–∑ –¥–∞—Ç –≤ —Ç–∞–±–ª–∏—Ü–∞—Ö —Å –∑–∞–∫–∞–∑–∞–º–∏ –∏ —Ä–∞—Å—Ö–æ–¥–∞–º–∏:

```python
orders_['order_month'] = orders_['buy_ts'].astype('datetime64[M]')
costs['month'] = costs['dt'].astype('datetime64[M]') 
```

–ü–æ—Å—á–∏—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–æ–≤—ã—Ö –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π (n_buyers) –∑–∞ –∫–∞–∂–¥—ã–π –º–µ—Å—è—Ü:

```python
first_orders = orders_.groupby('uid').agg({'order_month': 'min'}).reset_index()
first_orders.columns = ['uid', 'first_order_month']
cohort_sizes = (
    first_orders.groupby('first_order_month')
    .agg({'uid': 'nunique'})
    .reset_index()
)
cohort_sizes.columns = ['first_order_month', 'n_buyers']
cohort_sizes.head()
```

–ü–æ—Å—Ç—Ä–æ–∏–º –∫–æ–≥–æ—Ä—Ç—ã. –î–æ–±–∞–≤–∏–º –º–µ—Å—è—Ü –ø–µ—Ä–≤–æ–π –ø–æ–∫—É–ø–∫–∏ –∫–∞–∂–¥–æ–≥–æ –ø–æ–∫—É–ø–∞—Ç–µ–ª—è –≤ —Ç–∞–±–ª–∏—Ü—É —Å –∑–∞–∫–∞–∑–∞–º–∏:

```python
orders_ = pd.merge(orders_,first_orders, on='uid')
```

–°–≥—Ä—É–ø–ø–∏—Ä—É–µ–º —Ç–∞–±–ª–∏—Ü—É –∑–∞–∫–∞–∑–æ–≤ –ø–æ –º–µ—Å—è—Ü—É –ø–µ—Ä–≤–æ–π –ø–æ–∫—É–ø–∫–∏ –∏ –º–µ—Å—è—Ü—É –∫–∞–∂–¥–æ–≥–æ –∑–∞–∫–∞–∑–∞ –∏ —Å–ª–æ–∂–∏–º –≤—ã—Ä—É—á–∫—É. 

```python
cohorts = (
    orders_.groupby(['first_order_month', 'order_month'])
    .agg({'revenue': 'sum'})
    .reset_index()
)
cohorts.head()
```

–ö–∞–∂–¥–∞—è —Å—Ç—Ä–æ–∫–∞ —Ç–∞–±–ª–∏—Ü—ã cohorts –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç, –∫–∞–∫—É—é –≤—ã—Ä—É—á–∫—É –ø—Ä–∏–Ω–µ—Å–ª–∞ –∫–æ–≥–æ—Ä—Ç–∞.

–î–æ–±–∞–≤–∏–º –≤ —Ç–∞–±–ª–∏—Ü—É cohorts –¥–∞–Ω–Ω—ã–µ –æ —Ç–æ–º, —Å–∫–æ–ª—å–∫–æ –ª—é–¥–µ–π –ø–µ—Ä–≤—ã–π —Ä–∞–∑ —Å–æ–≤–µ—Ä—à–∏–ª–∏ –ø–æ–∫—É–ø–∫—É –≤ –∫–∞–∂–¥—ã–π –º–µ—Å—è—Ü:

```python
report_ltv = pd.merge(cohort_sizes, cohorts, on='first_order_month')
report_ltv.head()
```

–ü–æ—Å—á–∏—Ç–∞–µ–º –≤–∞–ª–æ–≤—É—é –ø—Ä–∏–±—ã–ª—å:

```python
# –º–∞—Ä–∂–∏–Ω–∞–ª—å–Ω–æ—Å—Ç—å —Å–µ—Ä–≤–∏—Å–∞
margin_rate = 1
# —Ä–∞—Å—á–µ—Ç –≤–∞–ª–æ–≤–æ–π –ø—Ä–∏–±—ã–ª–∏
report_ltv['gp'] = report_ltv['revenue'] * margin_rate
# —Ä–∞—Å—á–µ—Ç –≤–æ–∑—Ä–∞—Å—Ç–∞ –∫–æ–≥–æ—Ä—Ç—ã
report_ltv['age'] = ( report_ltv['order_month'] - report_ltv['first_order_month']) / np.timedelta64(1, 'M')
report_ltv['age'] = report_ltv['age'].round().astype('int')
report_ltv.head()
```

–†–∞–∑–¥–µ–ª–∏–º –≤–∞–ª–æ–≤—É—é –ø—Ä–∏–±—ã–ª—å –∫–æ–≥–æ—Ä—Ç –∑–∞ –∫–∞–∂–¥—ã–π –º–µ—Å—è—Ü –Ω–∞ –æ–±—â–µ–µ —á–∏—Å–ª–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤ –∫–∞–∂–¥–æ–π –∫–æ–≥–æ—Ä—Ç–µ. –î–ª—è –Ω–∞–≥–ª—è–¥–Ω–æ—Å—Ç–∏ –≤–∏–∑—É–∞–ª–∏–∑–∏—Ä—É–µ–º –∫–æ–≥–æ—Ä—Ç—ã –≤ –≤–∏–¥–µ —Å–≤–æ–¥–Ω–æ–π —Ç–∞–±–ª–∏—Ü—ã:

```python
# —Ä–∞—Å—á–µ—Ç ltv
report_ltv['ltv'] = report_ltv['gp'] / report_ltv['n_buyers']
output = report_ltv.pivot_table(
    index='first_order_month', columns='age', values='ltv', aggfunc='mean'
    ).cumsum(axis=1).round(2)

# –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ —Ç–µ–ø–ª–æ–≤–æ–π –∫–∞—Ä—Ç—ã
plt.figure(figsize=(15, 5))
plt.title('LTV')
sns.heatmap(output, annot=True, fmt=".1f", linewidths=.5, cmap = 'RdYlGn',vmin = 0.1, vmax = 6)
plt.xlabel('–í–æ–∑—Ä–∞—Å—Ç –∫–æ–≥–æ—Ä—Ç—ã, –º–µ—Å.')
plt.ylabel('–ö–æ–≥–æ—Ä—Ç–∞')
plt.show()
```

–ù–∞ —Ç–µ–ø–ª–æ–≤–æ–π –∫–∞—Ä—Ç–µ –º–æ–∂–Ω–æ —É–≤–∏–¥–µ—Ç—å, —Å–∫–æ–ª—å–∫–æ –≤ —Å—Ä–µ–¥–Ω–µ–º –∫–∞–∂–¥—ã–π –ø–æ–∫—É–ø–∞—Ç–µ–ª—å –∏–∑ –ø–µ—Ä–≤–æ–π –∫–æ–≥–æ—Ä—Ç—ã –ø—Ä–∏–Ω–µ—Å –ø—Ä–∏–±—ã–ª–∏. 
–ù–∞–ø—Ä–∏–º–µ—Ä, –≤ —Å—Ä–µ–¥–Ω–µ–º –∫–∞–∂–¥—ã–π –ø–æ–∫—É–ø–∞—Ç–µ–ª—å –∏–∑ –ø–µ—Ä–≤–æ–π –∫–æ–≥–æ—Ä—Ç—ã –ø—Ä–∏–Ω–µ—Å –ø–æ 11.9 —É.–µ. –≤–∞–ª–æ–≤–æ–π –ø—Ä–∏–±—ã–ª–∏ –∑–∞ 11 –º–µ—Å—è—Ü–µ–≤ "–∂–∏–∑–Ω–∏". 

–ë–æ–ª—å—à–µ –≤—Å–µ–≥–æ –ø—Ä–∏–±—ã–ª–∏ –ø—Ä–∏–Ω–µ—Å–ª–∞ —á–µ—Ç–≤–µ—Ä—Ç–∞—è –∫–æ–≥–æ—Ä—Ç–∞, –∑–∞ 8 –º–µ—Å—è—Ü–µ–≤ "–∂–∏–∑–Ω–∏" —Å—É–º–º–∞ —Å–æ—Å—Ç–∞–≤–∏–ª–∞ 13.4 —É.–µ. –ù–æ –≤ —ç—Ç–æ–π –∫–æ–≥–æ—Ä—Ç–µ –Ω–∞–±–ª—é–¥–∞–µ–º —Å—Ç—Ä–∞–Ω–Ω—ã–π –ø–∏–∫ –ø—Ä–æ–¥–∞–∂ –Ω–∞ 3-–π –º–µ—Å—è—Ü –∂–∏–∑–Ω–∏, –≤–æ–∑–º–æ–∂–Ω–æ —ç—Ç–æ –∞–Ω–æ–º–∞–ª—å–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –∏ —Å—Ç–æ–∏–ª–æ –±—ã –∏–∑ –∏–∑—É—á–∏—Ç—å –ø–æ–ø–æ–¥—Ä–æ–±–Ω–µ–µ, –Ω–æ —ç—Ç–æ –º–æ–∂–µ—Ç –∑–∞—Ç—è–Ω—É—Ç—å –∞–Ω–∞–ª–∏–∑, –ø–æ—ç—Ç–æ–º—É –ø–µ—Ä–µ–π–¥–µ–º –∫ —Å–ª–µ–¥—É—é—â–µ–º—É –ø—É–Ω–∫—Ç—É.

–†–∞—Å—Å—á–∏—Ç–∞–µ–º —Å—Ä–µ–¥–Ω–∏–π LTV –ø–æ –∫–æ–≥–æ—Ä—Ç–∞–º –∑–∞ 6 –º–µ—Å—è—Ü–µ–≤:

```python
m6_cum_ltv = output.mean(axis=0)[5]
print('–°—Ä–µ–¥–Ω–∏–π LTV –∑–∞ 6 –º–µ—Å—è—Ü–µ–≤ –ø–æ—Å–ª–µ –ø–µ—Ä–≤–æ–π –ø–æ–∫—É–ø–∫–∏: {:.2f}'.format(m6_cum_ltv))
```

```python
print('–í—Å–µ–≥–æ –ø–æ—Ç—Ä–∞—á–µ–Ω–æ –Ω–∞ –º–∞—Ä–∫–µ—Ç–∏–Ω–≥ {} —É.–µ.'.format(costs['costs'].sum()))
```

–í—ã—è—Å–Ω–∏–º, –∫–∞–∫ —Ç—Ä–∞—Ç—ã —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω—ã –ø–æ –∏—Å—Ç–æ—á–Ω–∏–∫–∞–º.

```python
# —Å–≥—Ä—É–ø–ø–∏—Ä—É–µ–º –¥–∞–Ω–Ω—ã–µ –ø–æ –∏—Å—Ç–æ—á–Ω–∏–∫—É –∏ –ø—Ä–æ—Å—É–º–º–∏—Ä—É–µ–º
costs_grouped_source = costs.groupby('source_id')['costs'].sum().reset_index()
# —Å–ø–∏—Å–æ–∫ –ø–æ—Ä—è–¥–∫–∞ –¥–ª—è –≤—ã–≤–æ–¥–∞ –∫–∞—Ç–µ–≥–æ—Ä–∏–π –∏—Å—Ç–æ—á–Ω–∏–∫–æ–≤ –Ω–∞ –æ—Å–∏ —Ö
x_order = costs_grouped_source['source_id'].to_list()
# –≥—Ä–∞—Ñ–∏–∫ —Å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –∏–≥—Ä –≤ —Ä–∞–∑–Ω—ã–µ –≥–æ–¥—ã
sns.catplot(
            kind="bar", data=costs_grouped_source,  x='source_id', y='costs', \
            alpha=.7, color='#ff0000', height=6, aspect=2, orient='v', order=x_order\
             )
# —É—Å—Ç–∞–Ω–æ–≤–∏–º –∑–∞–≥–æ–ª–æ–≤–æ–∫ –¥–ª—è –≥—Ä–∞—Ñ–∏–∫–∞
plt.title("–ó–∞—Ç—Ä–∞—Ç—ã –Ω–∞ —Ä–µ–∫–ª–∞–º—É –ø–æ –∏—Å—Ç–æ—á–Ω–∏–∫–∞–º"); 
# –ø–æ–¥–ø–∏—à–µ–º –æ—Å–∏ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç –æ–±–æ–∑–Ω–∞—á–∏–≤ –µ–¥–∏–Ω–∏—Ü—ã –∏–∑–º–µ—Ä–µ–Ω–∏—è
plt.ylabel('–ó–∞—Ç—Ä–∞—Ç—ã –Ω–∞ —Ä–µ–∫–ª–∞–º—É, —É.–µ.')
plt.xlabel('–ò—Å—Ç–æ—á–Ω–∏–∫–∏')
plt.show()
```

–ù–∞–∏–±–æ–ª—å—à–∏–µ –∑–∞—Ç—Ä–∞—Ç—ã –±—ã–ª–∏ –æ—Å—É—â–µ—Å—Ç–≤–ª–µ–Ω—ã –ø–æ –∏—Å—Ç–æ—á–Ω–∏–∫–∞–º 3(42.9%), 4(18,6%), 5(15,7%).

–†–∞—Å—Å–º–æ—Ç—Ä–∏–º –¥–∏–Ω–∞–º–∏–∫—É –∑–∞—Ç—Ä–∞—Ç –Ω–∞ —Ä–µ–∫–ª–∞–º—É –ø–æ –º–µ—Å—è—Ü–∞–º.

```python
costs['costs_month'] = costs['dt'].astype('datetime64[M]')

monthly_costs = costs.groupby('costs_month').agg({'costs':'sum'}).reset_index()
# –ø–æ–¥–ø–∏—Å–∏ –∫ –æ—Å–∏ –•, –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –º–µ—Å—è—Ü–∞ –≤ —Ñ–æ—Ä–º–∞—Ç %Y-%m
x_order = monthly_costs['costs_month'].dt.strftime('%Y-%m')
# –≥—Ä–∞—Ñ–∏–∫ –¥–∏–Ω–∞–º–∏–∫–∏ –∑–∞—Ç—Ä–∞—Ç –Ω–∞ —Ä–µ–∫–ª–∞–º—É
sns.catplot(
            kind="point", data=monthly_costs,  x='costs_month', y='costs', order = x_order,\
            alpha=.6, height=6, aspect=2\
            )
# —É—Å—Ç–∞–Ω–æ–≤–∏–º –∑–∞–≥–æ–ª–æ–≤–æ–∫ –¥–ª—è –≥—Ä–∞—Ñ–∏–∫–∞
plt.title("–î–∏–Ω–∞–º–∏–∫–∞ –∑–∞—Ç—Ä–∞—Ç –Ω–∞ —Ä–µ–∫–ª–∞–º—É –ø–æ –º–µ—Å—è—Ü–∞–º"); 
# –ø–æ–¥–ø–∏—à–µ–º –æ—Å–∏ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç –æ–±–æ–∑–Ω–∞—á–∏–≤ –µ–¥–∏–Ω–∏—Ü—ã –∏–∑–º–µ—Ä–µ–Ω–∏—è
plt.ylabel('–ó–∞—Ç—Ä–∞—Ç—ã, —É.–µ.')
plt.xlabel('–ú–µ—Å—è—Ü')
plt.xticks(rotation=30)
plt.show()
```

–¢–µ–ø–µ—Ä—å —Ä–∞—Å—Å–º–æ—Ç—Ä–∏–º –¥–∏–Ω–∞–º–∏–∫—É –∑–∞—Ç—Ä–∞—Ç –ø–æ –º–µ—Å—è—Ü–∞–º –≤ —Ä–∞–∑—Ä–µ–∑–µ –∏—Å—Ç–æ—á–Ω–∏–∫–æ–≤.

```python
# —Å–æ–∑–¥–∞–Ω–∏–µ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ —Å –∑–∞—Ç—Ä–∞—Ç–∞–º–∏ –≤ –∫–∞–∂–¥–æ–º –º–µ—Å—è—Ü–µ –ø–æ –∏—Å—Ç–æ—á–Ω–∏–∫–∞–º
monthly_costs_source = costs.groupby(['costs_month','source_id']).agg({'costs':'sum'}).reset_index()

# –ø–æ–¥–ø–∏—Å–∏ –∫ –æ—Å–∏ –•, –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –º–µ—Å—è—Ü–∞ –≤ —Ñ–æ—Ä–º–∞—Ç %Y-%m
monthly_costs_source['costs_month_Y_m'] = monthly_costs_source['costs_month'].dt.strftime('%Y-%m')
# –≥—Ä–∞—Ñ–∏–∫ –¥–∏–Ω–∞–º–∏–∫–∏ –∑–∞—Ç—Ä–∞—Ç –ø–æ –∏—Å—Ç–æ—á–Ω–∏–∫–∞–º
sns.catplot(
            kind="point", data=monthly_costs_source,  x='costs_month_Y_m', y='costs', hue="source_id", \
            alpha=.6, height=6, aspect=2\
            )
# —É—Å—Ç–∞–Ω–æ–≤–∏–º –∑–∞–≥–æ–ª–æ–≤–æ–∫ –¥–ª—è –≥—Ä–∞—Ñ–∏–∫–∞
plt.title("–ó–∞—Ç—Ä–∞—Ç—ã –ø–æ –º–µ—Å—è—Ü–∞–º –≤ —Ä–∞–∑—Ä–µ–∑–µ –∏—Å—Ç–æ—á–Ω–∏–∫–æ–≤"); 
# –ø–æ–¥–ø–∏—à–µ–º –æ—Å–∏ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç –æ–±–æ–∑–Ω–∞—á–∏–≤ –µ–¥–∏–Ω–∏—Ü—ã –∏–∑–º–µ—Ä–µ–Ω–∏—è
plt.ylabel('–ó–∞—Ç—Ä–∞—Ç—ã, —É.–µ.')
plt.xlabel('–ú–µ—Å—è—Ü–∞')
plt.xticks(rotation=30)
plt.show()
```

### –®–∞–≥ 3.3.2. –°—Ç–æ–∏–º–æ—Å—Ç—å –ø—Ä–∏–≤–ª–µ—á–µ–Ω–∏—è –∫–ª–∏–µ–Ω—Ç–∞

–†–∞—Å—Å—á–∏—Ç–∞–µ–º —Å—Ä–µ–¥–Ω–∏–π CAC –Ω–∞ –æ–¥–Ω–æ–≥–æ –ø–æ–∫—É–ø–∞—Ç–µ–ª—è –¥–ª—è –≤—Å–µ–≥–æ –ø—Ä–æ–µ–∫—Ç–∞ –∏ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –∏—Å—Ç–æ—á–Ω–∏–∫–∞ —Ç—Ä–∞—Ñ–∏–∫–∞.

CAC (–æ—Ç –∞–Ω–≥–ª. customer acquisition cost) ‚Äî —Å—Ç–æ–∏–º–æ—Å—Ç—å –ø—Ä–∏–≤–ª–µ—á–µ–Ω–∏—è –∫–ª–∏–µ–Ω—Ç–∞. –°—É–º–º–∞ –¥–µ–Ω–µ–≥, –≤–æ —Å–∫–æ–ª—å–∫–æ –æ–±—Ö–æ–¥–∏—Ç—Å—è –Ω–æ–≤—ã–π –∫–ª–∏–µ–Ω—Ç –∫–æ–º–ø–∞–Ω–∏–∏.

```python
# –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ —Ç–∞–±–ª–∏—Ü —Å —Ä–∞–∑–º–µ—Ä–∞–º–∏ –∫–æ–≥–æ—Ä—Ç—ã –∏ –∑–∞—Ç—Ä–∞—Ç–∞–º–∏ –Ω–∞ –Ω–∏—Ö –∫–∞–∂–¥—ã–π –º–µ—Å—è—Ü
# cohort_sizes - —Ä–∞–∑–º–µ—Ä –∫–æ–≥–æ—Ä—Ç—ã –ø–æ –º–µ—Å—è—Ü–∞–º
# monthly_costs - –∑–∞—Ç—Ä–∞—Ç—ã –Ω–∞ –∫–æ–≥–æ—Ä—Ç—É –ø–æ –º–µ—Å—è—Ü–∞–º
report_cac = pd.merge(cohort_sizes, monthly_costs, left_on='first_order_month', right_on='costs_month')
# —Ä–∞—Å—á–µ—Ç —Å—Ç–æ–∏–º–æ—Å—Ç–∏ –ø—Ä–∏–≤–ª–µ—á–µ–Ω–∏—è –æ–¥–Ω–æ–≥–æ –ø–æ–∫—É–ø–∞—Ç–µ–ª—è
report_cac['cac'] = report_cac.costs / report_cac.n_buyers

# –ø–æ–¥–ø–∏—Å–∏ –∫ –æ—Å–∏ –•, –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –º–µ—Å—è—Ü–∞ –≤ —Ñ–æ—Ä–º–∞—Ç %Y-%m
x_order = report_cac['costs_month'].dt.strftime('%Y-%m')
# –≥—Ä–∞—Ñ–∏–∫ —Å—Ç–æ–∏–º–æ—Å—Ç–∏ –ø—Ä–∏–≤–ª–µ—á–µ–Ω–∏—è –æ–¥–Ω–æ–≥–æ –ø–æ–∫—É–ø–∞—Ç–µ–ª—è
sns.catplot(
            kind="point", data=report_cac,  x='costs_month', y='cac', order = x_order, \
            alpha=.6, height=6, aspect=2\
            )
# —É—Å—Ç–∞–Ω–æ–≤–∏–º –∑–∞–≥–æ–ª–æ–≤–æ–∫ –¥–ª—è –≥—Ä–∞—Ñ–∏–∫–∞
plt.title("–î–∏–Ω–∞–º–∏–∫–∞ —Å—Ç–æ–∏–º–æ—Å—Ç–∏ –ø—Ä–∏–≤–ª–µ—á–µ–Ω–∏—è –æ–¥–Ω–æ–≥–æ –ø–æ–∫—É–ø–∞—Ç–µ–ª—è –ø–æ –º–µ—Å—è—Ü–∞–º"); 
# –ø–æ–¥–ø–∏—à–µ–º –æ—Å–∏ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç –æ–±–æ–∑–Ω–∞—á–∏–≤ –µ–¥–∏–Ω–∏—Ü—ã –∏–∑–º–µ—Ä–µ–Ω–∏—è
plt.ylabel('–ó–∞—Ç—Ä–∞—Ç—ã –Ω–∞ –ø—Ä–∏–≤–ª–µ—á–µ–Ω–∏–µ, —É.–µ.')
plt.xlabel('–ú–µ—Å—è—Ü')
plt.xticks(rotation=30)
plt.show()
```

–ù–∞ –≥—Ä–∞—Ñ–∏–∫–µ –≤–∏–¥–∏–º, —á—Ç–æ –≤ –∞–≤–≥—É—Å—Ç–µ –∑–∞—Ç—Ä–∞—Ç—ã –Ω–∞ –ø—Ä–∏–≤–ª–µ—á–µ–Ω–∏–µ –ø–æ–∫—É–ø–∞—Ç–µ–ª—è —Å–∏–ª—å–Ω–æ –≤–æ–∑—Ä–æ—Å–ª–∏, –∞ –∑–∞—Ç–µ–º –≤ —Å–µ–Ω—Ç—è–±—Ä–µ —Å–Ω–∏–∑–∏–ª–∏—Å—å –¥–æ —É—Ä–æ–≤–Ω—è –∏—é–ª—è, –≤–æ–∑–º–æ–∂–Ω–æ –ø—Ä–æ—à–ª–∞ –∫–∞–∫–∞—è-—Ç–æ —Ä–µ–∫–ª–∞–º–Ω–∞—è –∫–æ–º–ø–∞–Ω–∏—è. 

–ü–æ—Å—á–∏—Ç–∞–µ–º CAC –ø–æ –∏—Å—Ç–æ—á–Ω–∏–∫–∞–º, –≤ –∫–∞—á–µ—Å—Ç–≤–µ –∏—Å—Ç–æ—á–Ω–∏–∫–∞ –≤–æ–∑–º–µ–º —Ç–æ—Ç, —Å –∫–æ—Ç–æ—Ä–æ–≥–æ –∫–ª–∏–µ–Ω—Ç –ø—Ä–∏—à–µ–ª –≤–ø–µ—Ä–≤—ã–µ.

```python
# –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ –ø–µ—Ä–≤–æ–≥–æ –º–µ—Å—è—Ü–∞ –∏ –¥–∞—Ç—ã –ø–æ—Å–µ—â–µ–Ω–∏—è —Å–∞–π—Ç–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–º
visits['session_month'] = visits['start_ts'].astype('datetime64[M]')
visits['session_date'] = visits['start_ts'].dt.date

# —Ä–∞—Å—á–µ—Ç –¥–∞—Ç—ã –∏ –≤—Ä–µ–º–µ–Ω–∏ –ø–µ—Ä–≤–æ–π —Å–µ—Å—Å–∏–∏ –∫–∞–∂–¥–æ–≥–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è
first_visits = visits.groupby('uid').agg({'start_ts':'min'}).reset_index()
first_visits.columns = ['uid','first_session_start_ts']

first_visits['first_session_dt'] = first_visits['first_session_start_ts'].dt.date
first_visits['first_session_month'] = first_visits['first_session_start_ts'].astype('datetime64[M]')

# —Ä–∞—Å—á–µ—Ç –¥–∞—Ç—ã –ø–µ—Ä–≤–æ–π –ø–æ–∫—É–ø–∫–∏ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –∫–ª–∏–µ–Ω—Ç–∞
orders['buy_dt'] = orders['buy_ts'].dt.date
orders['order_month'] = orders['buy_ts'].astype('datetime64[M]')

# –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ –¥–∞—Ç—ã –ø–µ—Ä–≤–æ–π –ø–æ–∫—É–ø–∫–∏ –∫–∞–∂–¥–æ–≥–æ –ø–æ–∫—É–ø–∞—Ç–µ–ª—è
first_orders = orders.groupby('uid').agg({'buy_ts':'min'}).reset_index()
first_orders.columns = ['uid','first_order_ts']
first_orders['first_order_dt'] = first_orders['first_order_ts'].dt.date
first_orders['first_order_month'] = first_orders['first_order_ts'].astype('datetime64[M]')

# —Å–æ–∑–¥–∞–Ω–∏–µ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ —Å –∑–∞–∫–∞–∑–∞–º–∏ –∏ –∏—Å—Ç–æ—á–Ω–∏–∫–∞–º–∏
client = pd.merge(first_visits , first_orders, on='uid')

client.head()
```

```python
# —Å–æ–∑–¥–∞–Ω–∏–µ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ —Å –ø–µ—Ä–≤—ã–º —Ä–µ–∫–ª–∞–º–Ω—ã–º –∏—Å—Ç–æ—á–Ω–∏–∫–æ–º
users = visits.sort_values('start_ts').groupby('uid').first()
users = users[['source_id']]

# –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ –≤ —Ç–∞–±–ª–∏—Ü—É client –¥–∞–Ω–Ω—ã—Ö –æ–± –∏—Å—Ç–æ—á–Ω–∏–∫–∞—Ö
client = pd.merge(client, users, on = 'uid')
client['first_order_dt'] = pd.to_datetime(client['first_order_dt'])
client['first_session_dt'] = pd.to_datetime(client['first_session_dt'])
client.head()
```

```python
# –ø–æ–¥—Å—á–µ—Ç –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –ø–æ—Å–µ—Ç–∏—Ç–µ–ª–µ–π –≤ –¥–µ–Ω—å
client_day = client.groupby(['source_id','first_order_dt']).agg({'uid':'count'}).reset_index()
client_day.rename(columns={'uid':'n_client'}, inplace=True)
client_day['first_order_dt'] = pd.to_datetime(client_day['first_order_dt'])

client_day.head()
```

```python
# –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ —Ç–∞–±–ª–∏—Ü —Å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∫–ª–∏–µ–Ω—Ç–æ–≤ –≤ –¥–µ–Ω—å –∏ —Ä–∞—Å—Ö–æ–¥–∞–º–∏
report_cac = pd.merge(client_day, costs, left_on = ['source_id','first_order_dt'], right_on=['source_id','dt'])
# —Ä–∞—Å—á–µ—Ç –°–ê–°
report_cac['cac'] = report_cac['costs'] / report_cac['n_client']

result_cac = report_cac.pivot_table(
    index=['costs_month'],
    columns='source_id',
    values='cac',
    aggfunc='mean'
)
result_cac.plot(figsize = (16, 6))
# —É—Å—Ç–∞–Ω–æ–≤–∏–º –∑–∞–≥–æ–ª–æ–≤–æ–∫ –¥–ª—è –≥—Ä–∞—Ñ–∏–∫–∞
plt.title("–°—Ç–æ–∏–º–æ—Å—Ç—å –ø—Ä–∏–≤–ª–µ—á–µ–Ω–∏—è –æ–¥–Ω–æ–≥–æ –ø–æ–∫—É–ø–∞—Ç–µ–ª—è –ø–æ –∏—Å—Ç–æ—á–Ω–∏–∫–∞–º"); 
# –ø–æ–¥–ø–∏—à–µ–º –æ—Å–∏ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç –æ–±–æ–∑–Ω–∞—á–∏–≤ –µ–¥–∏–Ω–∏—Ü—ã –∏–∑–º–µ—Ä–µ–Ω–∏—è
plt.ylabel('–°—Ç–æ–∏–º–æ—Å—Ç—å –ø—Ä–∏–≤–ª–µ—á–µ–Ω–∏—è, —É.–µ.')
plt.xlabel('–ú–µ—Å—è—Ü')
plt.xticks(rotation=30)
plt.show()
```

–ò–∑ –≥—Ä–∞—Ñ–∏–∫–∞ –≤–∏–¥–Ω–æ, —á—Ç–æ —Å—Ç–æ–∏–º–æ—Å—Ç—å –ø—Ä–∏–≤–ª–µ—á–µ–Ω–∏—è –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π –∏–∑ —Ç—Ä–µ—Ç—å–µ–≥–æ –∏ –≤—Ç–æ—Ä–æ–≥–æ –∏—Å—Ç–æ—á–Ω–∏–∫–∞ —Å–∞–º–∞—è –≤—ã—Å–æ–∫–∞—è. –ü–æ—Å–ª–µ –∞–≤–≥—É—Å—Ç–∞ 2017–≥. —Å—Ç–æ–∏–º–æ—Å—Ç—å –ø—Ä–∏–≤–ª–µ—á–µ–Ω–∏—è –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π –∏–∑ —á–µ—Ç–≤–µ—Ä—Ç–æ–≥–æ –∏—Å—Ç–æ—á–Ω–∏–∫–∞ —Ä–µ–∑–∫–æ —Å–Ω–∏–∑–∏–ª–∞—Å—å. –ü–æ—Ö–æ–∂–µ–µ –ø–æ–≤–µ–¥–µ–Ω–∏–µ –µ—Å—Ç—å —Ç–∞–∫–∂–µ —É –¥–µ—Å—è—Ç–æ–≥–æ –∏—Å—Ç–æ—á–Ω–∏–∫–∞ –ø–æ—Å–ª–µ —è–Ω–≤–∞—Ä—è 2018–≥.

–í —Ü–µ–ª–æ–º –ø–æ –∏—Å—Ç–æ—á–Ω–∏–∫–∞–º 1,2,3,4,5,9 –Ω–∞–±–ª—é–¥–∞–µ—Ç—Å—è –Ω–µ–∫–∞—è —Ç–µ–Ω–¥–µ–Ω—Ü–∏—è —Å–Ω–∏–∂–µ–Ω–∏—è —Å—Ç–æ–∏–º–æ—Å—Ç–∏ –ø—Ä–∏–≤–ª–µ—á–µ–Ω–∏—è –ø–æ—Å–ª–µ –∞–≤–≥—É—Å—Ç–∞ 2017–≥.

–†–∞—Å—Å—á–∏—Ç–∞–µ–º —Å—Ä–µ–¥–Ω–∏–π –°–ê–° –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –∏—Å—Ç–æ—á–Ω–∏–∫–∞ —Ç—Ä–∞—Ñ–∏–∫–∞:

```python
# —Ä–∞—Å—á–µ—Ç —Å—Ä–µ–¥–Ω–µ–≥–æ –ø–æ–∫–∞–∑–∞—Ç–µ–ª—è CAC –ø–æ –≤—Å–µ–º –∏—Å—Ç–æ—á–Ω–∏–∫–∞–º
report_cac_mean = report_cac.groupby('source_id')['cac'].mean().reset_index()
print("–°—Ä–µ–¥–Ω–∏–π –°–ê–° –ø–æ –∫–∞–∂–¥–æ–º—É –∏—Å—Ç–æ—á–Ω–∏–∫—É —Ç—Ä–∞—Ñ–∏–∫–∞:\n",report_cac_mean)
```

### –®–∞–≥ 3.3.3. ROMI –ø–æ –∫–æ–≥–æ—Ä—Ç–∞–º –≤ —Ä–∞–∑—Ä–µ–∑–µ –∏—Å—Ç–æ—á–Ω–∏–∫–æ–≤

ROMI (–æ—Ç –∞–Ω–≥–ª. return on marketing investments, ¬´–≤–æ–∑–≤—Ä–∞—Ç –Ω–∞ –º–∞—Ä–∫–µ—Ç–∏–Ω–≥–æ–≤—ã–µ –∏–Ω–≤–µ—Å—Ç–∏—Ü–∏–∏¬ª). –†–µ–∫–ª–∞–º–∞ —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–∞, –µ—Å–ª–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ –ø—Ä–∏–Ω–æ—Å—è—Ç –±–æ–ª—å—à–µ –¥–µ–Ω–µ–≥, —á–µ–º —Ç—Ä–∞—Ç–∏—Ç –±–∏–∑–Ω–µ—Å –Ω–∞ –∏—Ö –ø—Ä–∏–≤–ª–µ—á–µ–Ω–∏–µ.

```python
### –ö–æ–¥ —Ä–µ–≤—å—é–µ—Ä–∞ –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –æ–±—â–µ–≥–æ ROMI.
report.pivot_table(index='first_order_month', columns='age', values='romi', aggfunc='mean').cumsum(axis=1)
```

<div class="alert alert-danger">
<b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞:</b> –ü—Ä–æ–≤–µ—Ä—å, –ø–æ–∂–∞–ª—É–π—Å—Ç–∞, —Ä–∞—Å—á–µ—Ç CAC. ROMI –Ω–∞ —Ñ–∏–Ω–∞–ª—å–Ω—ã—Ö –º–µ—Å—è—Ü–∞—Ö –∫–æ–≥–æ—Ä—Ç –ø–æ–ª—É—á–∏–ª—Å—è –∑–∞–≤—ã—à–µ–Ω –Ω–∞ 30 - 50%.
</div>

<div class="alert alert-info"> <b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Å—Ç—É–¥–µ–Ω—Ç–∞:</b> –ò—Å–ø—Ä–∞–≤–∏–ª–∞ –æ—à–∏–±–∫—É –≤ —Ä–∞—Å—á–µ—Ç–∞—Ö</div>

```python
def calculate_ltv(source_id):
    # —Ñ–∏–ª—å—Ç—Ä–∞—Ü–∏—è —Ç–∞–±–ª–∏—Ü—ã –ø–æ –∏—Å—Ç–æ—á–Ω–∏–∫—É  
    client_ = client.query('source_id == @source_id')
    # —Ä–∞—Å—á–µ—Ç ltv
    # —Ä–∞—Å—á–µ—Ç —á–∏—Å–ª–∞ –Ω–æ–≤—ã—Ö –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π –≤ –∫–∞–∂–¥—É—é –¥–∞—Ç—É
    cohorts_size_ = client_.groupby(['first_order_month']).agg({'uid': 'nunique'}).reset_index()
    cohorts_size_.rename(columns = {'uid':'n_clients'}, inplace=True)
    cohorts_ = pd.merge(orders , client_, how='inner', on = 'uid')\
            .groupby(['first_order_month','order_month'])\
            .agg({'revenue':'sum'}).reset_index()
    # —Ä–∞—Å—á–µ—Ç –≤–æ–∑—Ä–∞—Å—Ç–∞ –∫–æ–≥–æ—Ä—Ç—ã
    cohorts_['age_month'] = (
        (cohorts_['order_month'] - cohorts_['first_order_month']) / np.timedelta64(1, 'M')
    ).round().astype('int')

    report = pd.merge(cohorts_size_, cohorts_, on='first_order_month')
    # —Ä–∞—Å—á–µ—Ç ltv
    report['ltv'] = report.revenue / report.n_clients
    return report
```

```python
def calculate_romi(report, source_id):
    #—Ä–∞—Å—á–µ—Ç ROMI 
    report['romi']= report['ltv'] / report_cac_mean.loc[report_cac_mean.loc[:,'source_id'] == source_id]['cac'].iloc[0]

    result_romi = report.pivot_table(
        index=['first_order_month'],
        columns='age_month',
        values='romi',
        aggfunc='mean'
    ).cumsum(axis=1)
    # –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ —Ç–µ–ø–ª–æ–≤–æ–π –∫–∞—Ä—Ç—ã
    sns.set(style='whitegrid')
    # —Ä–∞–∑–º–µ—Ä —Ç–µ–ø–ª–æ–≤–æ–π –∫–∞—Ä—Ç—ã
    plt.figure(figsize=(13, 9))
    # –∑–∞–≥–æ–ª–æ–≤–æ–∫ —Ç–µ–ø–ª–æ–≤–æ–π –∫–∞—Ä—Ç—ã
    plt.title('ROMI –ø–æ –∏—Å—Ç–æ—á–Ω–∏–∫—É '+ str(source_id))
    # –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ —Ç–µ–ø–ª–æ–≤–æ–π –∫–∞—Ä—Ç—ã
    sns.heatmap(result_romi, annot=True, linewidths=1, linecolor='black',vmin = 0.01, vmax = 1);
    # –ø–æ–¥–ø–∏—à–µ–º –æ—Å–∏ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç –æ–±–æ–∑–Ω–∞—á–∏–≤ –µ–¥–∏–Ω–∏—Ü—ã –∏–∑–º–µ—Ä–µ–Ω–∏—è
    plt.ylabel('–ü–µ—Ä–≤—ã–π –º–µ—Å—è—Ü –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏')
    plt.xlabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –º–µ—Å—è—Ü–µ–≤ —Å –ø–µ—Ä–≤–æ–π –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏')   
#     print(result_romi.round(2).fillna(''))
```

```python
# —Ä–∞—Å—á–µ—Ç ltv –∏ romi –ø–æ –ø–µ—Ä–≤–æ–º—É –∏—Å—Ç–æ—á–Ω–∏–∫—É
ltv_1 = calculate_ltv(1)
romi_1 = calculate_romi(ltv_1,1)
```
