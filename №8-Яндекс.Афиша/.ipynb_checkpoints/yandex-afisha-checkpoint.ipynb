# Шаг 2. Подготовка данных  <a id="step2"></a>

Приведем названия столбцов таблиц visits и orders к нижнему регистру.

```python
# функция замены названий столбцов и приведения к нижнему регистру
def replace_names(data):
    list_of_columns = []
    for column in data.columns:
        column = column.lower()
        column = column.replace(' ', '_')
        list_of_columns.append(column)
    data.columns = list_of_columns
    return data
# заменим названия столбцов в таблице visits
visits = replace_names(visits)
# вывод названий столбцов
visits.columns
```

```python
# заменим названия столбцов в таблице orders
orders = replace_names(orders)
# вывод названий столбцов
orders.columns
```

Теперь названия столбцов таблиц visits и orders приведены к нижнему регистру и переименованы, пробелы в названиях заменены на нижнее подчеркивание, для удобства работы.

Далее изменим тип столбцов там где это требовалось.

```python
# изменение типа данных в столбцах
visits['device'] = visits['device'].astype('category')
print('Привели столбец device к типу',visits.device.dtypes) 

visits['start_ts'] = pd.to_datetime(visits['start_ts'])
visits['end_ts'] = pd.to_datetime(visits['end_ts'])
print('Привели столбец start_ts к типу',visits.start_ts.dtypes) 
print('Привели столбец end_ts к типу',visits.end_ts.dtypes) 

visits['source_id'] = visits['source_id'].astype('int8')
costs['source_id'] = costs['source_id'].astype('int8')
print('Привели столбец source_id к типу в таблицах visits и costs',visits.source_id.dtypes) 

orders['buy_ts'] = pd.to_datetime(orders['buy_ts'])
print('Привели столбец buy_ts к типу',orders.buy_ts.dtypes) 

costs['dt'] = pd.to_datetime(costs['dt'])
print('Привели столбец dt к типу',costs.dt.dtypes) 
```

# Шаг 3. Расчет метрик и построение графиков <a id="step3"></a>

## Шаг 3.1. Продуктовые метрики <a id="step4"></a>

Рассчитаем DAU, WAU и MAU. Вычислим средние значения этих метрик за весь период.

* DAU (от англ. daily active users) — количество уникальных пользователей в день;
* WAU (от англ. weekly active users) — количество уникальных пользователей в неделю;
* MAU (от англ. monthly active users) — количество уникальных пользователей в месяц.

Чтобы посчитать недельное и месячное использование, выделим в отдельные столбцы год, месяц и неделю, а также полную дату.

```python
# вычисление года, месяца, недели и даты из start_ts таблицы visits
visits['session_year']  = visits['start_ts'].astype('datetime64[Y]')
visits['session_month'] = visits['start_ts'].astype('datetime64[M]')
visits['session_week']  = visits['start_ts'].astype('datetime64[W]')
visits['session_date'] = visits['start_ts'].astype('datetime64[D]')
visits.sample(5) 
```

Сгруппируем данные по уникальным пользователям, найдём среднее и посчитаем метрики.

```python
dau_total = visits.groupby('session_date').agg({'uid': 'nunique'})
wau_total = visits.groupby(['session_year', 'session_week']).agg({'uid': 'nunique'})
mau_total = visits.groupby(['session_year','session_month']).agg({'uid':'nunique'})
print('В день сервисом в среднем пользуются {} людей'.format(int(dau_total.mean())))
print('В неделю сервисом в среднем пользуются {} людей'.format(int(wau_total.mean())))
print('В месяц сервисом в среднем пользуются {} людей'.format(int(mau_total.mean())))
```

```python
dau_total.plot.line(figsize = (16, 6), title = 'Среднее количество посещений Яндекс.Афиша в день')
plt.xlabel("Даты")
plt.ylabel("Количество посещений")
plt.show()
```

```python
wau_total.plot.line(figsize = (16, 6), title = 'Среднее количество посещений Яндекс.Афиша в неделю');
# подпишем оси координат обозначив единицы измерения
plt.xlabel("Недели")
plt.ylabel("Количество посещений")
plt.xticks(rotation=30)
plt.show()
```

```python
mau_total.plot.line(figsize = (16, 6), title = 'Среднее количество посещений Яндекс.Афиша в месяц');
plt.xlabel("Месяцы")
plt.ylabel("Количество посещений")
plt.xticks(rotation=30)
plt.show()
```

```python
sticky_wau = dau_total.mean() / wau_total.mean()
sticky_mau = dau_total.mean() / mau_total.mean()
print('Недельная вовлеченность аудитории: {:.1%}'.format(sticky_wau[0]))
print('Месячная вовлеченность аудитории: {:.1%}'.format(sticky_mau[0]))
```

Приличное число уникальных пользователей в месяц.

Мы узнали, сколько пользователей заходит на Яндекс.Афишу. Хорошо бы получить информацию, как долго они там сидят. Посморим, сколько раз за день пользователи в среднем заходят на сайт.

```python
sessions_per_user = visits.groupby(['session_year', 'session_month']).agg({'uid': ['count','nunique']})
sessions_per_user.columns = ['n_sessions', 'n_users']
sessions_per_user['sessions_per_user'] = sessions_per_user['n_sessions'] / sessions_per_user['n_users']
print(sessions_per_user) 
```

В месяц один пользователь в среднем проводит в приложении максимум 1 сессию. Не сказать, что продукт используется регулярно.

Посчитаем среднюю продолжительность сессии. Для начала посмотрим на график распределения:

```python
# добавление столбца со длительностью посещения сайта в секундах
visits['session_duration_sec'] = (visits['end_ts'] - visits['start_ts']).dt.seconds
```

```python
# гистограмма распределения 
sns.histplot(data=visits, x='session_duration_sec', color='#ff0000', alpha=.7, bins = 100, kde=True)
# установим заголовок для графика
plt.title("Гистограмма распределения длительности сессий")
# подпишем оси координат обозначив единицы измерения
plt.ylabel('Частота')
plt.xlabel('Длительность сессии, сек.')
```

Распределение не похоже на нормальное, поэтому рассчитаем моду (значение в выборке, которое встречается чаще всего):

```python
print("Мода выборки visits \n",visits['session_duration_sec'].mode()) 
```

Cредняя продолжительность сессии 60 секунд, но пока мы не можем оценить достаточно ли это для достижения конечной цели клиента(покупки билета). 

Рассчитаем Retention Rate(англ. «коэффициент удержания»), применяя когортный анализ. Определим событие и период — на их основе сформируем когорту. Возьмём дату, когда пользователь впервые проявил активность в сервисе Яндекс.Афиша.

```python
# расчет даты, когда пользователь проявил первую активность
first_activity = visits.groupby('uid').agg({'start_ts':'min'})
# # переименование столбца
first_activity.columns = ['first_activity']
try:
    # объединение таблиц visits и first_activity
    visits = visits.join(first_activity, on='uid')
    print("В таблицу visits добавлен столбец с датой первой активной сессии")
except:
    print("В таблице visits уже есть столбец с датой первой активной сессии")
```

Теперь для каждой строки датафрейма можно рассчитать lifetime пользователя в рамках когорты. 

```python
# добавление столбца first_activity_month, месяц первого посещения пользователя
visits['first_activity_month'] = visits['first_activity'].astype('datetime64[M]')
# добавление столбца start_ts_month, месяц начала сессии пользователя
visits['start_ts_month'] = visits['start_ts'].astype('datetime64[M]')
# расчет lifetime
visits['lifetime'] = ((visits['start_ts_month'] - visits['first_activity_month']) / np.timedelta64(1, 'M')).round().astype('int')
```

Сгруппируем данные по когорте и lifetime. Посчитаем для каждой когорты количество активных пользователей на определённый «месяц жизни»:

```python
cohorts = visits.groupby(['first_activity_month', 'lifetime']).agg({'uid':'nunique'}).reset_index()
initial_users_count = cohorts[cohorts['lifetime'] == 0][['first_activity_month', 'uid']]
initial_users_count.columns = ['first_activity_month', 'cohort_users']
initial_users_count = initial_users_count.set_index('first_activity_month')
```

Объединим данные по когортам с исходным количеством пользователей в когорте:

```python
cohorts = cohorts.merge(initial_users_count,on='first_activity_month') 
```

Наконец, рассчитаем Retention Rate. Разделим количество активных пользователей в каждую из месяцов на исходное число пользователей в когорте:

```python
cohorts['retention'] = cohorts['uid']/cohorts['cohort_users'] 
```

Построим сводную таблицу и создадим тепловую карту:

```python
retention_pivot = cohorts.pivot_table(index='first_activity_month',\
                                      columns='lifetime',\
                                      values='retention',\
                                      aggfunc='sum') 
```

Построим тепловую карту:

```python
sns.set(style='whitegrid')
# размер тепловой карты
plt.figure(figsize=(13, 9))
# заголовок тепловой карты
plt.title('Cohorts: User Retention')
# построение тепловой карты
sns.heatmap(retention_pivot, annot=True, fmt='0.1%', linewidths=1, linecolor='black', vmin = 0.01, vmax = 0.1);
# подпишем оси координат обозначив единицы измерения
plt.ylabel('Первый месяц активности')
plt.xlabel('Количество месяцев с первой активности')
```

В первой когорте пользователи удерживались лучше всего, на 12 месяце жизни когорты показатель 4,5% - это больше, чем на старте мартовской когорты.
 
Самый большой Retention Rate на первый месяц был у пользователей, пришедших в сентябре ~ 8,5%.

Больше всего своих пользователей потеряла когорта декабря, в которой к маю осталось 1.9%.

С каждой новой когортой поведение покупателей меняется, но в целом уровень возвратов пользователей на Яндекс.Афишу достаточно низкий. Наверное, потому что далеко не каждый человек посещает различные развлекательные мероприятия стабильно раз в месяц. Или в компании были выбраны неэффективные источники трафика или ухудшилось качество работы службы поддержки.

Найдем средний Retention Rate на второй месяц «жизни» когорт:

```python
try:
    # сброс мультииндекса
    retention_pivot = retention_pivot.reset_index()
    print("Мультииндекс сброшен")
except:
    print("Мультииндекс уже сброшен")
    
# расчет Retention Rate на второй месяц
retention_second_month = retention_pivot[2].sum() / retention_pivot[:10][0].sum()
print('Cредний Retention Rate на второй месяц «жизни» когорт = {:.5f}'.format(retention_second_month))
```

## Шаг 3.2. Метрики электронной коммерции <a id="step5"></a>

Исследуем, сколько времени в среднем проходит с момента первого посещения сайта до совершения покупки. Для этого найдем время первой покупки в таблице orders и добавим дату и время первой сессии в информацию по первым заказам. После вычислим разницу и узнаем количество времени, которое прошло до момента совершения первой покупки клиентом.

```python
# расчет даты и времени первой покупки клиента
first_orders = orders.groupby(['uid']).agg({'buy_ts':'min'})
# переименование столбца
first_orders.columns = ['first_orders']
try:
    # объединение таблиц orders и first_orders
    orders = orders.join(first_orders, on='uid')
    print("В таблицу orders добавлен столбец с датой первой покупки")
except:
    print("В таблице orders уже есть столбец с датой первой покупки")
```

Добавим в таблицу orders столбец с датой первой активной сессии, которую мы уже считали выше, данные хранятся в датафрейме first_activity, путем объединения таблиц orders и first_activity по ключу uid.

```python
try:
    # объединение таблиц orders и first_activity
    orders = orders.join(first_activity, on='uid')
    print("В таблицу orders добавлен столбец с датой первой активной сессии")
except:
    print("В таблице orders уже есть столбец с датой первой активной сессии")  
```

Найдем время, прошедщее с момента первой сессии до первой покупки.

```python
orders['time_to_first_purchase'] = (orders['first_orders'] - orders['first_activity']) / np.timedelta64(1, 'm')
print('Медиана времени с первой сессии до первой покупки: {} мин.'.format(int(orders['time_to_first_purchase'].median())))
```

```python
# размер графика
plt.figure(figsize=(15,5))
# гистограмма распределения 
orders['time_to_first_purchase'].plot(kind = 'hist', bins = 100, range = (0,50), color='#ff0000', alpha=.7)
# установим заголовок для графика
plt.title("Гистограмма распределения времени с момента первой сессии до первой покупки")
# подпишем оси координат обозначив единицы измерения
plt.ylabel('Частота')
plt.xlabel('Время до покупки, мин.')
```

Из графика наблюдаем, что большинство клиентов совершают первую покупку до 10 минут. Пик находится в раойне меньше 1 минуты. Медианное значение говорит нам о том, что время от первого посещения до совершения покупки около 22 минут. Выше мы получали, что средняя продолжительность сессии 60 секунд, за это время успевают сделать около 4000 покупок на сайте Яндекс.Афиша.

```python
# добавление столбца с днем совершения покупки
orders['buy_date'] = orders['buy_ts'].dt.date
# добавление столбца с неделей совершения покупки
orders['buy_week'] = orders['buy_ts'].dt.isocalendar().week
# добавление столбца с месяцом совершения покупки
orders['buy_month'] = orders['buy_ts'].dt.month
# добавление столбца с годом совершения покупки
orders['buy_year'] = orders['buy_ts'].dt.year

day_purchases_per_user = orders.groupby(['buy_date', 'uid']).agg({'buy_ts':'count'}).mean()[0]
week_purchases_per_user = orders.groupby(['buy_year', 'buy_week', 'uid']).agg({'buy_ts':'count'}).mean()[0]
month_purchases_per_user = orders.groupby(['buy_month', 'uid']).agg({'buy_ts':'count'}).mean()[0]
year_purchases_per_user = orders.groupby(['buy_year', 'uid']).agg({'buy_ts':'count'}).mean()[0]
print('В среднем на одного пользователя покупок в день {:.2f}'.format(day_purchases_per_user))
print('В среднем на одного пользователя покупок в месяц {:.2f}'.format(month_purchases_per_user))
print('В среднем на одного пользователя покупок в неделю {:.2f}'.format(week_purchases_per_user))
print('В среднем на одного пользователя покупок в год {:.2f}'.format(year_purchases_per_user))
```

Узнаем, как изменяется поведение каждой когорты на протяжении времени. Например, определим, снижается ли средний чек клиентов и одинаковый ли доход получает компания от клиентов, привлечённых в разные периоды.

Для начала сформируем когорты по месяцам покупки.

```python
# добавление столбца first_activity_month, месяц первого посещения пользователя
orders['first_activity_month'] = orders['first_activity'].astype('datetime64[M]')
# добавление столбца start_ts_month, месяц начала сессии пользователя
orders['buy_date'] = orders['buy_date'].astype('datetime64[M]')
# расчет lifetime
orders['lifetime'] = ((orders['buy_date'] - orders['first_activity_month']) / \
                      np.timedelta64(1, 'M')).round().astype('int')
cohorts = orders.groupby(['first_activity_month', 'lifetime'])\
                .agg({'revenue':'sum', 'uid':'nunique'})\
                .reset_index()
# переименование столбцов
cohorts.columns = ['first_activity_month', 'lifetime', 'revenue_total', 'n_users']
# выручка с одного пользователя
cohorts['revenue_per_user'] = cohorts['revenue_total'] / cohorts['n_users']
# когорты по сумме покупок на одного пользователя
cohorts_revenue_per_user = cohorts.pivot_table(index = 'first_activity_month',\
                                               columns = 'lifetime', \
                                               values = 'revenue_per_user', \
                                               aggfunc = 'mean')
# расчет среднего значения по всем месяцам жизни когорты
cohorts_revenue_per_user['mean'] = cohorts_revenue_per_user.mean(axis=1)
```

Построим тепловую карту:

Теперь посчитаем экономику одного покупателя.

LTV (от англ. lifetime value) — «пожизненная» ценность клиента. Это — итоговая сумма денег, которую один клиент приносит в среднем со всех своих покупок. При расчёте LTV обычно обращаются к валовой прибыли с клиента за период.

Маржинальность сервиса — 100%.

Посмотрим на распределение дат заказов:

```python
orders['first_orders'].describe() 
```

Первая покупка состоялась 1 июня 2017 года. Последняя — спустя ровно год 1 июня 2018 года.

```python
costs['dt'].describe() 
```

Диапазон дат расходов совпадает с датами заказов.

LTV удобно считать по когортам. Для этого нужно:
* найти дату первой покупки каждого покупателя;
* вычислить число новых покупателей за каждую дату;
* добавить дату первой покупки в таблицу с заказами;
* сгруппировать таблицу и посчитать выручку.

Выделим месяцы из дат в таблицах с заказами и расходами:

```python
orders['order_month'] = orders['buy_ts'].astype('datetime64[M]')
costs['month'] = costs['dt'].astype('datetime64[M]') 
```

Посчитаем количество новых покупателей (n_buyers) за каждый месяц:

```python
first_orders = orders.groupby('uid').agg({'order_month': 'min'}).reset_index()
first_orders.columns = ['uid', 'first_order_month']
cohort_sizes = (
    first_orders.groupby('first_order_month')
    .agg({'uid': 'nunique'})
    .reset_index()
)
cohort_sizes.columns = ['first_order_month', 'n_buyers']
cohort_sizes.head()
```

Построим когорты. Добавим месяц первой покупки каждого покупателя в таблицу с заказами:

```python
orders = pd.merge(orders,first_orders, on='uid')
```

Сгруппируем таблицу заказов по месяцу первой покупки и месяцу каждого заказа и сложим выручку. 

```python
cohorts = (
    orders.groupby(['first_order_month', 'order_month'])
    .agg({'revenue': 'sum'})
    .reset_index()
)
cohorts.head()
```

Каждая строка таблицы cohorts показывает, какую выручку принесла когорта.

Добавим в таблицу cohorts данные о том, сколько людей первый раз совершили покупку в каждый месяц:

```python
report_ltv = pd.merge(cohort_sizes, cohorts, on='first_order_month')
report_ltv.head()
```

Посчитаем валовую прибыль:

```python
# маржинальность сервиса
margin_rate = 1
# расчет валовой прибыли
report_ltv['gp'] = report_ltv['revenue'] * margin_rate
# расчет возраста когорты
report_ltv['age'] = ( report_ltv['order_month'] - report_ltv['first_order_month']) / np.timedelta64(1, 'M')
report_ltv['age'] = report_ltv['age'].round().astype('int')
report_ltv.head()
```

Разделим валовую прибыль когорт за каждый месяц на общее число пользователей в каждой когорте. Для наглядности визуализируем когорты в виде сводной таблицы:

```python
# расчет ltv
report_ltv['ltv'] = report_ltv['gp'] / report_ltv['n_buyers']
output = report_ltv.pivot_table(
    index='first_order_month', columns='age', values='ltv', aggfunc='mean'
    ).round(2)

# сумма ltv по когорте
output['sum'] = output.sum(axis=1)

# построение тепловой карты
plt.figure(figsize=(15, 5))
plt.title('LTV')
sns.heatmap(output, annot=True, fmt=".1f", linewidths=.5, cmap = 'RdYlGn',vmin = 0.1, vmax = 6)
plt.xlabel('Возраст когорты, мес.')
plt.ylabel('Когорта')
plt.show()
```

На тепловой карте можно увидеть, сколько в среднем каждый покупатель из первой когорты принес прибыли. 
Например, в среднем каждый покупатель из первой когорты принес по 11.9 у.е. валовой прибыли за 11 месяцев "жизни". 

Больше всего прибыли принесла четвертая когорта, за 8 месяцев "жизни" сумма составила 13.4 у.е. Но в этой когорте наблюдаем странный пик продаж на 3-й месяц жизни, возможно это аномальные данные и стоило бы из изучить поподробнее, но это может затянуть анализ, поэтому перейдем к следующему пункту.

Рассчитаем средний LTV по когортам за 6 месяцев:

```python
m6_cum_ltv = output.cumsum(axis=1).mean(axis=0)[5]
print('Средний LTV за 6 месяцев после первой покупки: {:.2f}'.format(m6_cum_ltv))
```

Чтобы вложения в маркетинг окупились, привлечение одного покупателя не должно превышать 7.97 у.е.

CAC (от англ. customer acquisition cost) — стоимость привлечения клиента. Сумма денег, во сколько обходится новый клиент компании.
